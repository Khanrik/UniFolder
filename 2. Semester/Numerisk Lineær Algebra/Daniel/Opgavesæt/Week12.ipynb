# 12.1



For en given egenværdi $\lambda$ og egenvektor $v$, så har vi følgende relation for en kvadratisk matrix $A$:

$$Av = \lambda v$$

Dette kan omskrives til:

$$Av - \lambda v = Av - \lambda I_n v = (A-\lambda I_n)v = 0$$

Vi ved at $v \neq 0$, da dette er en del af definitionen for en egenvektor. Vi ved dertil også at hvis matricen er invertibel, så er dens nulrum blot $\{0\}$. Dermed må vi have at denne ligning kun holder hvis $A-\lambda I_n$ er ikke-invertibel.

Dette er ensbetydende med at $\det(A-\lambda I_n) = 0$.

For vores matrice $A$ løser vi for $\lambda$ i følgende ligning:

$$\det(A-\lambda I_n) = \det(\begin{bmatrix} 7 - \lambda & 1 \\ 6 & 2 - \lambda \end{bmatrix})$$
$$= (7-\lambda)(2-\lambda)-1*6 = 14 - 7\lambda + \lambda^2 - 2\lambda - 6 = \lambda^2 - 9 \lambda + 8$$
$$= (\lambda - 8)(\lambda - 1) = 0$$

Løsningerne på denne ligning er tydelige vha $0$-reglen:

$\lambda_1 = 8$ og $\lambda_2 = 1$

Indsættes disse to egenværdier i vores matrix $A-\lambda I_n$, så skulle vi gerne få to lineært afhængige søjler, og i dette tilfælde er de også parallele.

### $\lambda_1 = 8$

$$A - 8 i_2 = \begin{bmatrix} 7 - 8 & 1 \\ 6 & 2 - 8 \end{bmatrix} = \begin{bmatrix} -1 & 1 \\ 6 & -6 \end{bmatrix}$$

Dette reduceres ved elementære rækkeoperationer (vores resultat vektor behøver ikke blive sat på til højre, da det blot er $0$-vektoren):

$$\rightarrow \begin{bmatrix} -1 & 1 \\ 0 & 0 \end{bmatrix}$$

Vi ved at denne matrix ganget på vores vektor $v = \begin{bmatrix} x \\ y \end{bmatrix}$ gerne skulle give vektoren $\begin{bmatrix} 0 \\ 0 \end{bmatrix}$

Vi har nu altså ligningssystemet:

$$\begin{bmatrix} -1 & 1 \\ 0 & 0 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$

Ud fra denne har vi to ligninger:

$$-x + y = 0$$ 
og
$$0 + 0 = 0$$

Her kan vi blot vælge en vilkårlig værdi for $x$, fx $x = 1$, og derfra finde en værdi for $y$:

$$-1 + y = 0 \Leftrightarrow y = 1$$

Dermed må den tilhørende egenvektor til $\lambda_1 = 8$ være $v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$

Det samme gøres nu for vores anden egenværdi $\lambda_2 = 1$:

### $\lambda_2 = 1$

$$A - 8 i_2 = \begin{bmatrix} 7 - 1 & 1 \\ 6 & 2 - 1 \end{bmatrix} = \begin{bmatrix} 6 & 1 \\ 6 & 1 \end{bmatrix}$$
$$\rightarrow \begin{bmatrix} 6 & 1 \\ 0 & 0 \end{bmatrix}$$

$$6x + y = 0$$

Lad $x = 1$:

$$6 + y = 0 \Leftrightarrow y = -6$$

Dermed må den tilhørende egenvektor til $\lambda_2 = 1$ være $v_2 = \begin{bmatrix} 1 \\ -6 \end{bmatrix}$

En vigtig note er at egenvektorer ikke er entydigt bestemte, de kan et hvert multiplum af den fundne egenvektor.

I dette tilfælde kunne $v_2$ også være:

$$v_2 = \begin{bmatrix} 2 \\ -12 \end{bmatrix}$$

Vi kan nu bestemme det karakteristiske polynomium af matricen $A$ på følgende måde:

$$\det(A - \lambda I_n) = \det(\begin{bmatrix} 6 - \lambda & -3 & 2 \\ 0 & 2 - \lambda & 7 \\ 0 & 0 & -1 - \lambda \end{bmatrix})$$
$$= (6-\lambda)((2-\lambda)(-1-\lambda) - 0 * 7) - (-3)(0 * -1 - 0 * 7) + 2(0 * 0 - (2-\lambda) * 0) =(6-\lambda)(2-\lambda)(-1-\lambda) = 0$$

## Redegørelse for at egenværdierne i en øvre triangulær matrix netop er diagonal indgangene:

Vi vil vise at $A = \begin{bmatrix} \lambda_1 & * & \dots & * \\
                                    0 & \lambda_2 & \dots & * \\
                                    \vdots & \vdots & \ddots & \vdots \\
                                    0 & 0 & \dots & \lambda_n   \end{bmatrix}$

Her vil vi ved at $\lambda$ og $v$ skal opfylde: $Av = \lambda v \Leftrightarrow (A-I_n\lambda) v = 0$

Vi kan nu bestemme egenværdierne ved at løse ligningen $\det(A - \lambda I_n) = 0$, fordi $A$ skal være ikke-invertibel.

Her er det ret tydeligt at vi af de sidste 2 rækker kan lave en masse 2x2 matricer. Af disse 2x2 matricer er det kun den sidste der har en determinant forskellig fra $0$. Denne vil få en faktor ganget på fra sin "parent" matrice $n-2$ gange, og her ser vi at vi får et polynomium opskrevet på form af sine rødder. Alle disse rødder vil være diagonalerne.

Dermed må diagonalerne netop være egenværdierne.

Vi ved også at $\det(A) = \det(A^T)$. Hvis dette antages, så er det meget nemt at vise.

## Er en sådan matrix nødvendigvis diagonaliserbar

Nej, hvis der er en gentaget egenværdi (Samme værdi på to diagonale indgange), så kan vi have at denne egenværdi blot har 1 egenvektor, som dermed ikke udgør en basis. Et eksempel på netop dette ses i eksempel 21.10 i notesættet.

## Opfylder en generelt øvretriangulær matrix en ligning af formen:

$$\Pi_{i=0}^{n-1} (A-\lambda_i I_n) = 0$$

Her er $A = \begin{bmatrix} \lambda_0 & * & \dots & * \\ 0 & \lambda_1 & \dots & * \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & \lambda_{n-1} \end{bmatrix}$

Ja, dette kan vises vha induktion.

Hver matrix $(A-\lambda_i I_n)$ er en øvretriangulær matrix med EN 0-diagonalindgang.

Hver matrix $(A - \lambda_i I_n)$ kummuterer fordi $\lambda_i$ kan have en hvilken som helst rækkefølge.

Gang dem sammen i følgende rækkefølge:

$$(A - \lambda_0 I_n)(A - \lambda_1 I_n)(A - \lambda_2 I_n)\dots$$

Kig på de to første 
$$(A - \lambda_0 I_n) = \begin{bmatrix} 0 & * & \dots & * \\ 0 & * & \dots & * \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & \lambda_{n-1} \end{bmatrix}$$
$$(A - \lambda_1 I_n) = \begin{bmatrix} * & * & \dots & * \\ 0 & 0 & \dots & * \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & \lambda_{n-1} \end{bmatrix}$$

Vi kan betragte et matrix produkt $AB$ som $\begin{bmatrix} Ab_0 & Ab_1 & \dots \end{bmatrix}$:

Læg her mærke til at den første søjle i $(A-\lambda_0 I_n)$ er en nul-vektor. Derudover er alle indgange, på nær den første, i den første søjle i $(A - \lambda_1 I_n)$. Dette vil sige at vores første søjle bliver en nulvektor.

Dette vil også gælde for den 2. søjle i vores produkt, af samme årsag.

Med induktion kan dette nu køres videre.

```python
A = np.array([[6, -3, 2, 1, 7],
              [0, 2, 7, 2 , 4],
              [0, 0, -1, 3, 5],
              [0, 0, 0, 5, 2],
              [0, 0, 0, 0, 1]], dtype=float)

A1 = A-6*np.eye(5)
A2 = A-2*np.eye(5)
A3 = A+np.eye(5)
A4 = A-5*np.eye(5)
A5 = A-np.eye(5)


A1, A2, A3, A4, A5
```

```python
A1, A1 @ A2, A1 @ A2 @ A3, A1 @ A2 @ A3 @ A4, A1 @ A2 @ A3 @ A4 @ A5
```
