## Problem 2
Let a second-order differential equation be given as
$$
x'' + \beta x' + 2x = 2 \cos(t)
$$

### 1. Solve the differential equation assuming $\beta = 3$, $x(0) = 1$, and $x'(0) = 0$. Plot the solution for $t \geq 0$.

To solve the differential equation, the general solution has to be found. With the given value of $\beta = 3$, the auxiliary of the complementary homogeneous equation becomes
$$
r^{2} + 3r + 2 = 0
$$

This expression can be factorized as follows
$$
(r + 1) (r + 2) = 0
$$

which means that it has the roots $r = -1$ and $r = -2$, giving the complementary equation
$$
x_{c} = c_{1} e^{-t} + c_{2}e^{-2t}
$$

Since the right hand side of the original equation is a sinosoidal function, it can be reasoned that a particular solution to the differential equation would be of the form
$$
x_{p} = B \cos(t) + C \sin(t)
$$

The derivatives of this particular solution would be
$$
x_{p}' = -B \sin(t) + C \cos(t)
$$
$$
x_{p}'' = -B \cos(t) - C \sin(t)
$$

Inserting these expressions into the original differential equation gives
\begin{align*}
-B \cos(t) - C \sin(t) + 3 (-B \sin(t) + C \cos(t)) + 2(B \cos(t) + C \sin(t)) &= 2 \cos(t)\\
-B \cos(t) - C \sin(t) - 3B \sin(t) + 3C \cos(t) + 2B \cos(t) + 2C \sin(t) &= 2 \cos(t)\\
B \cos(t) + 3C \cos(t) + C \sin(t) - 3B \sin(t) &= 2 \cos(t)\\
(B + 3C) \cos(t) + (C - 3B) \sin(t) &= 2\cos(t)
\end{align*}

This will be true for all $t$ provided that the cosine and sine terms on the two sides of this equation agree. It therefore sufﬁces for A and B to satisfy the two linear equations
\begin{align*}
B + 3C &= 2\\
C - 3B &= 0
\end{align*}

Solving this by substitution gives
\begin{align*}
C &= 3B &\text{(isolating C in equation 2)}\\
B + 3 (3B) &= 2 &\text{(substituting into equation 2)}\\
7B &= 2\\
B &= \frac{2}{7}\\
C &= - 3 \cdot \frac{2}{7} &\text{(substituting back into C)}\\
C &= - \frac{6}{7}
\end{align*}

This means that the general solution for the differential equation and its derivative is
\begin{align*}
x(t) &= c_{1} e^{-t} + c_{2}e^{-2t} + \frac{2}{7} \cos(t) - \frac{6}{7} \sin(t)\\
x'(t) &= -c_{1}e^{-t} - 2c_{2}e^{-2t} - \frac{2}{7} \sin(t) - \frac{6}{7} \cos(t)
\end{align*}

Inserting the inital value for the general solution gives
\begin{align*}
c_{1} e^{-0} + c_{2}e^{-2 \cdot 0} + \frac{2}{7} \cos(0) - \frac{6}{7} \sin(0) &= 1\\
c_{1} + c_{2} + \frac{2}{7} + 0 &= 1\\
c_{1} &= 1 - \frac{2}{7} - c_{2}\\
c_{1} &= \frac{5}{7} - c_{2}
\end{align*}

Inserting the found value of $c_{1}$ and the inital value into the derivative gives
\begin{align*}
-\left( \frac{5}{7} - c_{2} \right) e^{-0} - 2c_{2}e^{-2 \cdot 0} - \frac{2}{7} \sin(0) - \frac{6}{7} \cos(0) &= 0\\
-\frac{5}{7} + c_{2} - 2c_{2} - 0 - \frac{6}{7} &= 0\\
-c_{2} - \frac{11}{7} &= 0\\
c_{2} &= - \frac{11}{7}
\end{align*}

Inserting back into $c_{1}$ gives
$$
c_{1} = \frac{13}{7} - (- \frac{11}{7}) = \frac{24}{7}
$$

The particular solution of the differential equation then becomes
$$
x(t) = \frac{24}{7} e^{-t} - \frac{11}{7}e^{-2t} - \frac{6}{7} \cos(t) + \frac{2}{7} \sin(t)
$$
