## Problem *3.14

Given a causal system described by $y[n] = \frac{1}{2} y[n − 1] + x[n]$, compute its response to the following inputs:

### (a) $x[n] = e^{ j(\frac{\pi}{4}) n}, \quad -\infty < n < \infty$

Taking the z-transform leads to
$$
\begin{align*}
Y(z) &= \frac{1}{2} z^{-1} Y(z) + X(z)\\
Y(z) - \frac{1}{2} z^{-1} Y(z) &= X(z)\\
Y(z) \left( 1 - \frac{1}{2} z^{-1} \right) &= X(z)\\
Y(z) &= \frac{X(z)}{1 - \frac{1}{2} z^{-1}}\\
H(z) = \frac{Y(z)}{X(z)} &= \frac{1}{1 - \frac{1}{2} z^{-1}}
\end{align*}
$$

For LTI-systems, the response can be found as $y[n] = H[z] \cdot z^{n}$, where $x[n] = z^{n}$
$$
y[n] = \frac{1}{1 - \frac{1}{2} e^{ -j(\frac{\pi}{4})}} \cdot e^{ j(\frac{\pi}{4}) n}
$$

### (b) $x[n] = e^{ j(\frac{\pi}{4}) n} u[n]$

Finding the z-transform of $x[n]$ leads to
$$
X(z) = \frac{1}{1 - e^{ j\left(\frac{\pi}{4}\right) } z^{-1}}
$$

Using the same frequency response we can find $Y(z)$ using by rewriting equation 3.79 as $Y(z) = X(z) H(z)$.

\begin{align*}
Y(z) &= \frac{1}{1 - e^{ j\left(\frac{\pi}{4}\right) } z^{-1}} \frac{1}{1 - \frac{1}{2} z^{-1}}\\
&= \frac{A}{1 - e^{ j\left( \frac{\pi}{4} \right) } z^{-1}} + \frac{B}{1 - \frac{1}{2} z^{-1}}
\end{align*}

We use partial fractions to get easier z-transforms. Finding the partial fraction leads to:

\begin{align*}
\frac{1}{\left( 1 - e^{ j\left( \frac{\pi}{4} \right) } z^{-1} \right) \cdot \left( 1 - \frac{1}{2} z^{-1} \right) } &= \frac{A}{1 - e^{ j\left( \frac{\pi}{4} \right) } z^{-1}} + \frac{B}{1 - \frac{1}{2} z^{-1}}\\
1 &= A \left( 1 - \frac{1}{2} z^{-1} \right) + B \left( 1 - e^{ j\left( \frac{\pi}{4} \right) } z^{-1} \right)\\
&= A + B - \left(A \frac{1}{2} + B e^{ j \left(\frac{\pi}{4}\right)}\right) z^{-1}
\end{align*}

For this to hold for all $z$ it has to be true that

$$
1 = A + B \land 0 = A \frac{1}{2} + B e^{ j (\frac{\pi}{4}) }
$$

$$
B = 1 - A
$$

\begin{align*}
0 &= A \frac{1}{2} + (1 - A) e^{ j \left( \frac{\pi}{4} \right) }\\
-e^{ j \frac{\pi}{4} } &= A \left( \frac{1}{2} - e^{ j \frac{\pi}{4} } \right)\\
A &= -\frac{e^{ j \frac{\pi}{4} }}{\frac{1}{2} - e^{ j \frac{\pi}{4} }}\\
&= -\frac{2e^{ j \frac{\pi}{4} }}{1 - 2e^{ j \frac{\pi}{4} }}\\
&= \frac{1}{1 - 2e^{ -j \frac{\pi}{4} }}
\end{align*}

\begin{align*}
B &= 1 - \frac{1}{1 - 2e^{ -j \frac{\pi}{4} }}\\
&= \frac{1 - 2e^{ -j \frac{\pi}{4} } - 1}{1 - 2e^{ -j \frac{\pi}{4} }}\\
&= \frac{- 2e^{ -j \frac{\pi}{4} }}{1 - 2e^{ -j \frac{\pi}{4} }}\\
&= \frac{1}{1 - 2e^{ j \frac{\pi}{4} }}
\end{align*}

Doing an inverse z-transform on $Y(z)$ and inserting these values into the expression gives:

\begin{align*}
y[n] &= A e^{ j \frac{\pi}{4} n } u[n] + B \left( \frac{1}{2} \right)^{n} u[n]\\
&= \frac{e^{ j \frac{\pi}{4} n }}{1 - 2e^{ -j \frac{\pi}{4} }} u[n] + \frac{1}{2^{n} - 2^{n+1}e^{ j \frac{\pi}{4} }} u[n]
\end{align*}

### (c) $x[n] = (-1)^{n}, \quad -\infty < n < \infty$

The same frequency response the output response can again be found as $y[n] = H[z] \cdot z^{n}$, where $x[n] = z^{n}$

\begin{align*}
y[n] &= \frac{1}{1 - \frac{1}{2} (-1)^{ -1 }} \cdot (-1)^{ n } \\
&= \frac{1}{\frac{3}{2}} \cdot (-1)^{ n } \\
&= \frac{2}{3} \cdot (-1)^{ n }
\end{align*}

### (d) $x[n] = (-1)^{n} u[n]$

Finding the z-transform of $x[n]$ leads to
$$
X(z) = \frac{1}{1 + z^{-1}}
$$

Using the same frequency response we can find $Y(z)$ using by rewriting equation 3.79 as $Y(z) = X(z) H(z)$.

\begin{align*}
Y(z) &= \frac{1}{1 + z^{-1}} \frac{1}{1 - \frac{1}{2} z^{-1}} \\
&= \frac{A}{1 + z^{-1}} + \frac{B}{1 - \frac{1}{2} z^{-1}}
\end{align*}

We use partial fractions to get easier z-transforms. Finding the partial fraction leads to:

\begin{align*}
\frac{1}{\left( 1 + z^{-1} \right) \cdot \left( 1 - \frac{1}{2} z^{-1} \right) } &= \frac{A}{1 + z^{-1}} + \frac{B}{1 - \frac{1}{2} z^{-1}} \\
1 &= A (1 - \frac{1}{2} z^{-1}) + B (1 + z^{-1}) \\
1 &= A + B - (A \frac{1}{2} - B) z^{-1}
\end{align*}

For this to hold for all $z$ it has to be true that

$$
1 = A + B \land 0 = A \frac{1}{2} - B
$$

$$
B = 1 - A
$$

\begin{align*}
0 &= A \frac{1}{2} - (1 - A) \\
0 &= A \frac{3}{2} - 1 \\
A &= \frac{2}{3}
\end{align*}

\begin{align*}
B &= 1 - \frac{2}{3}\\
&= \frac{1}{3}
\end{align*}

Doing an inverse z-transform on $Y(z)$ and inserting these values into the expression gives:

\begin{align*}
y[n] &= A (-1)^{ n } u[n] + B \left( \frac{1}{2} \right)^{n} u[n]\\
&= \frac{2}{3} (-1)^{n} u[n] + \frac{1}{3} \left( \frac{1}{2} \right)^{n} u[n]
\end{align*}
