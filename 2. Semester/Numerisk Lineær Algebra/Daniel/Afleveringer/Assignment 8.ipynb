Bestemmelse af det karakteristiske polynomium:

$det(A-\lambda I_3) = det(\begin{bmatrix} 3-\lambda & 0 & 0 \\ 0 & 1 -\lambda & 1 \\ 0 & 1 & 1 - \lambda \end{bmatrix}) = (3-\lambda) * det(\begin{bmatrix} 1 - \lambda & 1 \\ 1 & 1 - \lambda \end{bmatrix}) - 0 * det(\begin{bmatrix} 0 & 1 \\ 0 & 1 - \lambda \end{bmatrix}) + 0 * det(\begin{bmatrix} 0 & 1 - \lambda \\ 0 & 1 \end{bmatrix})$

$= (3-\lambda) * det(\begin{bmatrix} 1 - \lambda & 1 \\ 1 & 1 - \lambda \end{bmatrix}) = (3-\lambda) * ((1 - \lambda)*(1 - \lambda) - 1*1) = (3-\lambda) * (\lambda^2 + 1 - 2\lambda - 1) = (3-\lambda) * (\lambda^2 - 2\lambda)$

$= - \lambda^3 + 5\lambda^2 - 6\lambda$

## f)

$$A^k = V \Lambda^k V^{-1}$$

Dette vil jeg nu bevise ved induktion:

Jeg sætter basecase til at være $k = 0$:

$$A^0 = V \Lambda^0 V^{-1}$$

Venstre:

$$A^0 = I_n$$

Højre:

Da V's søjler er lineært uafhængige, så må den være invertibel, og dermed må følgende holde:

$$V \Lambda^0 V^{-1} = VV^{-1} = I_n$$

Da ligheden er sand, så må sætningen altså holde for vores basecase.


Jeg vil nu bevise at sætningen holder for et tal $k+1$, under antagelsen at sætningen holder for tallet $k$ (IH: = $A^k = V \Lambda^k V^{-1}$):

$$A^{k+1} = V \Lambda^{k+1} V^{-1}$$

$$\parallel$$

$$A^k A = V \Lambda^k \Lambda V^{-1}$$

Jeg kan nu gange $V^{-1} V$ ind i midten, da det svarer til at gange med identitetsmatricen.

$$A^k A = V \Lambda^k V^{-1} V \Lambda V^{-1}$$

Da A er diagonaliserbar, så ved vi at $A = V \Lambda V^{-1}$, derudover ved har vi også vores antagelse ($A^k = V \Lambda^k V^{-1}$). I følge disse to må ligheden ovenfor være sand. Dermed må sætningen nedenfor gælde for alle $k \geq 0$.

$$A^k = V \Lambda^k V^{-1}$$



Dermed må sætningen nedenfor gælde for alle $k$, hvis $k \in \mathbb{N}$.

$$A^k = V \Lambda^k V^{-1}$$
$$\square$$
