## (1) What is the value of the following expressions? {-}

### 1. {-}
```ocaml
let x = 1 in
let f y = x in
let x = 2 in
f 0
```

Since OCaml utilises static binding, the closest defined value of x is used in the function f. Since `x = 2` is not yet declared when the function is defined, it will use `x = 1`. The expression will then have the value
```
int = 1
```

### 2. {-}
```ocaml
let f x = x - 1 in
let f x = f (x - 1) in
f 2
```

The second definition of f can be seen as $f(x-1) = (x - 1) - 1 = x = 2$, since youre substituting $x - 1$ into x for the first definition of f. When taking $f(2)$ the expression will then have the value
```
int = 0
```

### 3. {-}
```ocaml
let y = 2 in
let f x = x + y in
let f x = let y = 3 in f y in
f 5
```

The function is first defined as $f(x) = x + y$, where $y = 2$. It is then defined as $f(x) = y + 2$, where $y = 3$. Regardless of the input, the expression will then have the value
```
int = 5
```

### 4. {-}
```ocaml
let rec factorial i =
    if i = 0 then 1 else i * factorial (i - 1)
in
    factorial 5
```

This is a siple recursive function, which has the base case where if the input is 0 then the value is 1, and the recursive step where the current input is multiplied with $f(i-1)$. This definition would not catch negative numbers, but since the input is 5 then it will have the value
```
int = 120
```

## (2) Write a function sum that, given two integer bounds n and m and a function f, computes a summation. {-}
$$
\tag{1}
sum \; n \; m \; f = \sum_{i = n}^{m} f(i).
$$

```ocaml
let sum n m f = 
    let result = ref 0 in 
    for i = n to m do
        result := !result + f i
    done ;
    !result ;;
```

## (3) What is the value of the following expressions? {-}

### 1. {-}
```ocaml
let x = ref 1 in let y = x in y := 2; !x
```

y can be seen as a pointer to x, so when the value of what y points to is set to 2 (dereferenced by using :=), then x will get that value. The expression will then have the value
```
int = 2
```

### 2. {-}
```ocaml 
let x = ref 1 in let y = ref 1 in y := 2
```

Since nothing is banged then the expression will have the value
```
unit = ()
```

$\pagebreak$

### 3. {-}
```ocaml
let x = ref 1 in
let y = ref x in
!y := 2;
!x
```

Here y is a pointer pointer to x. Therefore it is dereferenced by banging, and the value pointed to (x) is set to 2. The expression will then have the value
```
int = 2
```

## (4) In the lectures we looked at the following imperative implementation of the factorial function: {-}
```ocaml
let factorial i =
let j = ref 1 in
for k = 2 to i do
    j := !j * k
done ;
```
Rewrite this function using:

### (a) The “downto” for loop construct. {-}

```ocaml
let factorial i =
    let j = ref 1 in
    for k = i downto 2 do
        j := !j * k
    done ;
```

### (b) The “while” for loop construct {-}

```ocaml
let factorial i =
    let j = ref 1 in
    let k = ref 2 in
    while !k <= i do
        j := !j * !k;
        k := !k + 1
    done ;
```

## (5) Consider the line equation: {-}
$$
\tag{2}
y = m * x + c

### (a) Create a function "*line*" that takes three *labelled* input arguments (m,x,c) and returns the corresponding *y value*. {-}

```ocaml
let line m x c = m *. x +. c
```

```ocaml
let line m x c = m *. x +. c in
let ramp x = line 0.45 x 10.
```

### (c) Print out the "y" values of my ramp for x = 0 to x = 1 in increments of 0.1 metre. {-}

```ocaml
let line m x c = m *. x +. c in
let ramp x = line 0.45 x 10. in
let i = ref 0. in
while !i <= 1. do
    Printf.printf "x=%f: %f \n" !i (ramp !i);
    i := !i +. 0.1
done;
```
