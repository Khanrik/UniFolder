## (1) For each of the following expressions, is the expression well-typed? If it is well-typed, does it evaluate to a value? If so, what is the value? {-}

### 1. 1 - 2 {-}

Both operands are integers so it is well-typed. It will have the value
```
int = -1
```

### 2. 1 - 2 - 3 {-}

All operands are integers so it is well-typed. It will have the value
```
int = -4
```

### 3. 1 - - 2 {-}

Double minus would just lead to plus, making this well-typed. It will have the value
```
int = 3
```

### 4. 0b101 - 0x10 {-}

Integers can be denoted in decimal, octal, binary, and hexadecimal, making this well-typed. Since `0b101=5` and `0x10=16` It will have the value
```
int = -11
```

### 5. 1073741823 + 1 {-}

Assuming your integers are at least 32 bit, then both are integers which can simply be added. It will have the value
```
int = 1073741824
```

### 6. 1073741823.0 + 1e2 {-}

Both of the operands are floats, but the operator is the integer addition and not float addition (+.). This is therefore not well-typed.

### 7. 1 ^ 1 {-}

The operands are not string and can not be used with the string concatenation operation. This is therefore not well-typed.

### 8. if true then 1 {-}

Here there is an implicit else branch which evaluates to `()`. Every expression has has to have one type, which makes this invalid. This is therefore not well-typed.

### 9. if false then () {-}

This expression will always have a value of type unit making it well-typed. It will have the value
```
unit = ()
```

### 10. if 0.3 -. 0.2 = 0.1 then 'a' else 'b' {-}

This is all valid since the expressions are both characters. The condition is mathematically correct, but because of floating point precision it won't be true, there it will have the value
```
char = 'b'
```

### 11. true || (1 / 0 >= 0) {-}

`||` is the disjunction/or operation which shortcircuits, circumventing the division by zero and making it well-typed. It will have the value
```
bool = true
```

### 12. 1 > 2 - 1 {-}

This is a simple comparison between two integers and is well-typed. It will have the value
```
bool = false
```

### 13. "Hello world”.[6] {-}

Strings support random access making this well-typed. It will have the value
```
char = 'w'
```

### 14. "Hello world”.[11] <- ’s’ {-}

In the newer version of OCaml strings have become immutable, but according to the book this should be valid syntax. Assuming we follow the book, the new string would be `string = "Hello worls"`, but the expression would return
```
unit = ()
```

### 15. String.lowercase "A" < "B" {-}

This is also deprecated in newer version of OCaml, but using String.lower_ascii instead would make it into a comparison between "a" (ASCII 97) and "B" (ASCII 66). It will have the value
```
bool = false
```

### 16. Char.code 'a' {-}

This is well-typed and would return the ASCII code of 'a'. It will have the value
```
int = 97
```

### 17. (((()))) {-}

Nesting unit elements is valid making this well typed. It will have the value
```
unit = ()
```

### 18. ((((\*1\*)))) {-}

This just inserts the commen `(*1*)` into the nested unit elements, which doesn't change anything. It will have the value

```
unit = ()
```

### 19. ((\*((()\*)) {-}
This simply places the comment `(*((()*)` into a `()`. It will have the value
```
unit = ()
```

## (2) Which of the following let-expressions is legal? For each expression that is legal, give its type and the value that it evaluates to. Otherwise, explain why the expression is not legal. {-}

### 1. let x = 1 in x {-}

This defines the identifier x as the value 1 and has the body x. This is legal and will evaluate to
```
int = 1
```

### 2. let x = 1 in let y = x in y {-}

This defines the identifier x as the value 1, then the identifier y as the value of x, and has the body y. This is legal and will evaluate to
```
int = 1
```

### 3. let x = 1 and y = x in y {-}

The `and` keyword lets you connect several `let` expressions. Since x is not defined in the second let definition (the first x is not in scope) it is not bound, which makes this expression illegal.

### 4. let x = 1 and x = 2 in x {-}

In this example x is defined in both `let` expressions, which is also illegal.

### 5. let x = 1 in let x = x in x {-}

This defines the identifier x as the value 1, then redefines x as the value of x, and has the body x. This is legal and will evaluate to
```
int = 1
```

### 6. let a' = 1 in a' + 1 {-}

This simply defines variable with the identifier a'. This is legal since it starts with lowercase letter or underscore, and only has legal characters. It will evaluate to
```
int = 2
```

### 7. let 'a = 1 in 'a + 1 {-}

This defines a variable which has an identifier that does not start with a lowercase letter or underscore, which makes it illegal.

## (3) What is the value of the following expression: {-}
```ocaml
let x = 1 in let x = x + 1 in x
```

This expression defines x as 1 and then as the value of x + 1, which results in 2. The body of the expression is x, which makes it evaluate to
```
int = 2
```
