## (3) Consider the follow OCaml function that we discussed in Lecture 4:
```pl
let rec even_rec x =
    if x = 0
    then true
    else odd_rec (x - 1)
and odd_rec x =
    if x = 0
    then false
    else even_rec (x - 1)
```

### 1. Start a new Prolog plain text file called “even.pl”, and implement these two functions as two Prolog predicates: even/1, odd/1
*Hint*: Look at how we implemented the function `numc` in the lecture slides with
respect to integer arithmetic in Prolog. Pay close attention to how we ensured that
the given argument was a natural number in `numc`.


The functions are implemented as follows
```pl
even(X) :-
    X = 0;
    X > 0,
    X0 is X - 1,
    odd(X0);
odd(X) :-
    X > 0,
    X0 is X - 1,
    even(X0);
```

### 2. Run the following queries, and record whether they are true or false:
```pl
?- even(0).
?- even(1).
?- even(2).
?- even(3).

?- odd(0).
?- odd(1).
?- odd(2).
?- odd(3).
```

The queries are runs as shown and returns the following
```
?- even(0).
true .

?- even(1).
false.

?- even(2).
true .

?- even(3).
false.

?- odd(0).
false.

?- odd(1).
true .

?- odd(2).
false.

?- odd(3).
true .
```

### 3. Add a new rule for `even/1` and `odd/1` to handle negative integer arguments. Run the following queries and record whether Prolog gives *true* or *false*:
```
?- even(-1).
?- even(-2).
?- even(-3).

?- odd(-1).
?- odd(-2).
?- odd(-3).
```

The code is now
```pl
even(X) :-
    X = 0;
    X > 0,
    X0 is X - 1,
    odd(X0);
    X < 0,
    X0 is X + 1,
    odd(X0).
odd(X) :-
    X > 0,
    X0 is X - 1,
    even(X0);
    X < 0,
    X0 is X + 1,
    even(X0).
```

And running the queries gives
```
?- even(-1).
false.

?- even(-2).
true .

?- even(-3).
false.

?- odd(-1).
true .

?- odd(-2).
false.

?- odd(-3).
true .
```

### 2. Implement a predicate less than or equal with the indicator leq/2 such that it is true when applied to two symbolic natural numbers X, Y when X is less than or equal to Y. Here are some test queries:
```
?- leq(z,z).
true.
?- leq(suc(z),z).
false.
?- leq(suc(suc(z)), suc(suc(suc(z))) ).
true.
```

The predicate is implemented as follows and works for the given queries:
```pl
leq(z, Y).
leq(suc(X), suc(Y)) :- leq(X, Y).
```

### 3. Consider the following OCaml function that we discussed in Lecture 4 for calculating the product of two symbolic natural numbers:
```pl
let rec mul_rec_nat x y =
match y with
Z -> Z
| Suc w -> add_rec_nat x (mul_rec_nat x w)
```
Implement this function as a predicate mul/3. Run the following queries and record the output:
```pl
?- mul(z, z, z).
?- mul(X, suc(z), Y).
?- numc(2,X), numc(3,Y), mul(X,Y,N).
```

It is implemented as follows:
```pl
mul(X, z, z).
mul(X, suc(Y), Z) :- mul(X, Y, W), add(X, W, Z).
```

And running the queries results in
```
?- mul(z, z, z).
true.

?- mul(X, suc(z), Y).
X = Y.

?- numc(2,X), numc(3,Y), mul(X,Y,N).
X = suc(suc(z)),
Y = suc(suc(suc(z))),
N = suc(suc(suc(suc(suc(suc(z)))))) 
```
