### 2. Et andet signal er givet ved $x_2[n] = 3x[n]$. Hvordan er $h[n]*x[n]$ og $h[n]*x_2[n]$ relateret?

Foldning er givet ved følgende formel (2.40)
$$
h[n]*x[n] = \sum_{k=-\infty}^{\infty} x[k] h[n-k]
$$

Indsættes $x_2[n] = 3x[n]$ fås
$$
\sum_{k=-\infty}^{\infty} x_2[k] h[n-k] = \sum_{k=-\infty}^{\infty} 3 x[k] h[n-k]
$$

Da konstanter ganget på kan trækkes udenfor summen kan det ses at foldningen med $x_2[n]$ bare er 3 gange større
$$
h[n]*x_2[n] = 3 \sum_{k=-\infty}^{\infty} x[k] h[n-k] = 3 (h[n]*x[n])
$$

## Opgave 2
Antag, at et filter har systemfunktionen
$$
H(z) = \frac{3 + z^{-1}}{1 + 0.5 \cdot z^{-1}}
$$

### 1. Bestem differensligningen for filteret.

\begin{align*}
H(z) = \frac{Y(z)}{X(z)} &= \frac{3 + z^{-1}}{1 + 0.5 z^{-1}}\\
Y(z) (1 + 0.5 z^{-1}) &= X(z) (3 + z^{-1})\\
Y(z) + 0.5 Y(z) z^{-1} &= 3X(z) + X(z) z^{-1}\\
Y(z) &= 3X(z) + X(z) z^{-1} - 0.5 Y(z) z^{-1}
&\Updownarrow\\
y[n] &= 3x[n] + x[n-1] -0.5 y[n-1]
\end{align*}

### 2. Tegn det tilhørende blokdiagram eller signal flow graph for filteret og angiv hvilken form, der er vist på tegningen.

Signal flow graph for filteret er vist forneden i den normale (ikke transponeret) første direkte form.



### 3. Skitser frekvenskarakteristikken af filteret.

```python
num = [3, 1]
den = [1, 0.5]

w = np.linspace(-np.pi, np.pi, 1000)
_, H = signal.freqz(num, den, worN=w)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(w / np.pi, 20 * np.log10(np.abs(H)))
plt.title('Magnitude Respons')
plt.xlabel('Normaliseret Frekvens ($\pi$ rad/sample)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(w / np.pi, np.unwrap(np.angle(H)))
plt.title('Fase Respons')
plt.xlabel('Normaliseret Frekvens ($\pi$ rad/sample)')
plt.ylabel('Fase (rad)')
plt.grid(True)

plt.show()
```
