```python
import numpy as np
import matplotlib.pyplot as plt
```

Betragt matricen
$$
A =
\begin{bmatrix}
3 & 0 & 0 \\
0 & 1 & 1 \\
0 & 1 & 1
\end{bmatrix}
$$

## (a) Bestem det karakteristiske polynomium $\text{det} (A -λ  I_{3})$ af $𝐴$. {-}

Det karakteristiske polynomium sættes op som vist i afsnit 21.5 af notesættet.
\begin{align*}
\text{det}(A - \lambda I_{3}) & = \text{det}\left( \begin{bmatrix}
3 - \lambda & 0 & 0 \\
0 & 1 - \lambda & 1 \\
0 & 1 & 1 - \lambda
\end{bmatrix} \right)\\
 & = (3 - \lambda) \text{det} \left(\begin{bmatrix}
1 - \lambda & 1 \\
1 & 1 - \lambda
\end{bmatrix}\right) - 0 \, \text{det} \left(\begin{bmatrix}
0 & 1 \\
0 & 1 - \lambda
\end{bmatrix}\right) + 0 \, \text{det} \left(\begin{bmatrix}
0 & 1 - \lambda \\
0 & 1
\end{bmatrix}\right)\\
 & = (3 - \lambda) (\lambda^{2} - 2\lambda) - 0 + 0\\
 & = 3 \lambda^{2} - 6 \lambda - \lambda^{3} + 2 \lambda^{2}\\
 & = - \lambda^{3} + 5 \lambda^{2} - 6 \lambda
\end{align*}

## (b) Vis at 0 og 3 er egenværdier for 𝐴. {-}

Proposition 21.15 fra notesættet siger at $\lambda$ er en egenværdi, hvis $\text{det} (A - \lambda I_{3}) = 0$. Værdierne kan da indsættes i det karakteristiske polynomium fra før, hvor det kan tjekkes om de giver 0.

$$
- 0^{3} + 5 \cdot 0^{2} - 6 \cdot 0 = 0 + 0 - 0 = 0
$$
$$
- 3^{3} + 5 \cdot 3^{2} - 6 \cdot 3 = - 27 + 45 - 18 = 0
$$

Da begge værdier resulterer i 0, må de begge være egenværdier for $A$.

For at se alle egenværdier afbildes funktionen 

```python
x = np.linspace(-1, 4, 100)

fig, ax = plt.subplots()
ax.plot(x, -(x**3) + 5*(x**2) - 6*x)
ax.grid(axis='x')
ax.axhline(0, color='black')
```

- 2^{3} + 5 \cdot 2^{2} - 6 \cdot 2 = - 8 + 20 - 12 = 0

## (d) Bestem en basis af $\mathbb{R}^{3}$, som består af egenvektorer for $A$. {-}

Først findes egenvektorerne for $A$ ved at løse ligningssystemet $(A - \lambda I_{3}) v = 0$ for hver egenværdi.

For $\lambda = 0$
$$
A - \lambda_{0} I_{3} =
\begin{bmatrix}
3 - 0 & 0 & 0 \\
0 & 1 - 0 & 1 \\
0 & 1 & 1 - 0
\end{bmatrix}
=
\begin{bmatrix}
3 & 0 & 0 \\
0 & 1 & 1 \\
0 & 1 & 1
\end{bmatrix}
\sim_{R_{2} \to R_{2} - R_{1}}
\begin{bmatrix}
3 & 0 & 0 \\
0 & 1 & 1 \\
0 & 0 & 0
\end{bmatrix}
$$
Følgende ligninger skal løses
$$
\begin{cases}
3x = 0 \\
y + z = 0
\end{cases}
$$
En løsning kan være
$$
v_{0} = 
\begin{bmatrix}
0 \\
1 \\
-1
\end{bmatrix}
$$

For $\lambda = 2$
$$
A - \lambda_{1} I_{3} =
\begin{bmatrix}
3 - 2 & 0 & 0 \\
0 & 1 - 2 & 1 \\
0 & 1 & 1 - 2
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 \\
0 & - 1 & 1 \\
0 & 1 & - 1
\end{bmatrix}
\sim_{R_{2} \to R_{2} + R_{1}}
\begin{bmatrix}
1 & 0 & 0 \\
0 & - 1 & 1 \\
0 & 0 & 0
\end{bmatrix}
$$
Følgende ligninger skal løses
$$
\begin{cases}
x = 0 \\
- y + z = 0
\end{cases}
$$
En løsning kan være
$$
v_{1} = 
\begin{bmatrix}
0 \\
1 \\
1
\end{bmatrix}
$$

For $\lambda = 3$
$$
A - \lambda_{2} I_{3} =
\begin{bmatrix}
3 - 3 & 0 & 0 \\
0 & 1 - 3 & 1 \\
0 & 1 & 1 - 3
\end{bmatrix}
=
\begin{bmatrix}
0 & 0 & 0 \\
0 & - 2 & 1 \\
0 & 1 & - 2
\end{bmatrix}
\sim_{R_{2} \to R_{2} + 2R_{1}}
\begin{bmatrix}
0 & 0 & 0 \\
0 & - 2 & 1 \\
0 & - 4 & 0
\end{bmatrix}
$$
Følgende ligninger skal løses
$$
\begin{cases}
-2 y + z = 0 \\
-4 y = 0
\end{cases}
$$
En løsning kan være
$$
v_{2} = 
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

Ifølge proposition 21.5 af notesættet vil egenvektorerne være en basis, hvis mængden af egenvektorer og -værdier svarer til matricens dimensioner. Ud fra de fundne egenvektorer kan en basis af $\mathbb{R}^{3}$ være:
$$
\begin{bmatrix}
0 \\
1 \\
-1
\end{bmatrix}
,
\begin{bmatrix}
0 \\
1 \\
1
\end{bmatrix}
,
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

## (e) Skriv $A = V Λ V^{-1}$ for passende $V$ og diagonal $Λ$. {-}

Som vist i proposition 21.6 af notesættet defineres $V$ som en samling af egenvektorer for $A$, og $\Lambda$ defineres som en diagonalmatrix bestående af egenværdierne. Udtrykket kan da opskrives for at tjekke om det går op (gøres med numpy.all()).

```python
A = np.array([  [3, 0, 0],
                [0, 1, 1],
                [0, 1, 1]])

V = np.array([  [0, 0, 1],
                [1, 1, 0],
                [-1, 1, 0]])

Lambda = np.diag(np.array([0, 2, 3]))

np.all(A == V @ Lambda @ np.linalg.inv(V))
```

## (f) Brug del (e) til at bestemme $A^{k}$ for alle $k$. {-}

Ud fra udtrykket fra del (e) vides det at $A^{k}$ kan skrives således:
$$
A^{k} = (V \Lambda V^{-1})^{k}
$$
Ses der eksempeltvis på $k=2$ ville det svare til følgende:
$$
A^{2} = (V \Lambda V^{-1}) (V \Lambda V^{-1}) = V \Lambda I_{2} \Lambda V^{-1} = V \Lambda^{2} V^{-1}
$$
Det kan ses at for hver faktor der ganges på med potensen vil $V^{-1}$ gå ud med $V$, hvilket efterlader en $\Lambda$ med $k$ som potens. For alle $k$ vil $A^{k}$ da svare til:
$$
A^{k} = V \lambda^{k} V^{-1}
$$
