## Opgave 4.1

Går med følgende defintioner:
$$
A \in \mathbb{R}^{m \times n}, \; B,C \in \mathbb{R}^{n \times r}, \; D \in \mathbb{R}^{r \times m}
$$

```python
m = 1000
n = 70
r = 3000
A = np.ones((m,n))
B, C = np.ones((n,r)), np.ones((n,r))
D = np.ones((r, m))
```

```python
def product_flops(M1,M2):
    S1 = M1.shape
    S2 = M2.shape
    return S1[0] * S2[1] * 2 * S2[0]
```

```python
def sum_flops(M1,M2):
    S1 = M1.shape
    return S1[0] * S1[1]
```

### (a)

```python
flop_amount = product_flops(A,B)

print(f"AB kræver {flop_amount} flops")
```

### (b)

```python
flop_amount = product_flops(A, B+C) + sum_flops(B,C)

print(f"A(B + C) kræver {flop_amount} flops")
```

### (c)

```python
flop_amount = product_flops(C, D) + product_flops(C@D, A)

print(f"(CD)A kræver {flop_amount} flops")
```

### (d)

```python
flop_amount = product_flops(D, A) + product_flops(C, D@A)

print(f"C(DA) kræver {flop_amount} flops")
```

### (e)

```python
BD = B @ D
flop_amount = product_flops(B, D) + product_flops(A, BD) + product_flops(A@BD, A)

print(f"(A(BD))A kræver {flop_amount} flops")
```

### (f)

```python
AB = A @ B
DA = D @ A
flop_amount = product_flops(A, B) + product_flops(D, A) + product_flops(AB, DA)

print(f"(AB)(DA) kræver {flop_amount} flops")
```

### (g)

```python
DA = D @ A
flop_amount = product_flops(D, A) + product_flops(B, DA) + product_flops(A, B@DA)

print(f"A(B(DA)) kræver {flop_amount} flops")
```
