```python
def IP71(u, v):
    w = np.array([7, 2, 3]) [:, np.newaxis]
    utemp = u * w
    return np.vdot(utemp, v)
```

```python
u = np.array([1, -1, 1]) [:, np.newaxis]
v = np.array([-1, -2, 1]) [:, np.newaxis]

IP71(u, v)
```

## (a)

Frobenius-normen:



Det tilhørende indre produkt:



```python
A = np.array([[6, 0],
              [0, -3]], dtype=float)

B = np.array([[1, 2],
              [-1, -2]], dtype=float)
```

```python
# Definition af det tilhørende indre produkt for Frobeniusnormen
def FrobeniusVdot(A, B):
    m, n = A.shape
    sum = 0
    for i in range(m):
        for j in range(n):
            sum += A[i, j] * B[i, j]
    return sum
    
```

```python
# Frobeniusnormen af A
np.linalg.norm(A, ord = 'fro'), np.sqrt(FrobeniusVdot(A, A))
```

```python
# Frobeniusnormen af B
np.linalg.norm(B, ord = 'fro'), np.sqrt(FrobeniusVdot(B, B))
```

```python
FrobeniusVdot(A, B)
```

```python
AFnorm = np.linalg.norm(A, ord = 'fro')
BFnorm = np.linalg.norm(B, ord = 'fro')
ABFdot = FrobeniusVdot(A, B)

print(ABFdot, '<=', AFnorm*BFnorm, 'is', ABFdot <= AFnorm*BFnorm)
```

```python
thetaRad = np.arccos(ABFdot / (AFnorm * BFnorm))
thetaDegree = thetaRad / (2 * np.pi) * 360

thetaRad, thetaDegree
```

# 7.5



I en QR-dekomponering har vi to matricer: $Q$ og $R$

$Q$ er en matrix som indeholder en ortonormal samling af vektorer som søjler, hvor den 0. søjle blot er enhedsvektoren for 0. søjle i $A$.

$R$ er en øvre triangulær matrix, som indeholder projektionerne af de andre vektorer på den pågældende vektor.

## (a)

$$A = \begin{bmatrix} 2 & 0 \\ 0 & 1 \end{bmatrix}$$

$u_0 = \begin{bmatrix} 2 \\ 0 \end{bmatrix}$

$w_0 = u_0$

$r_{00} = ||w_0||_2 = 2$

$v_0 = \frac{w_0}{||w_0||_2} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$



$u_1 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$

$w_1 = u_1 - proj_{v_0}(u_1) = \begin{bmatrix} 0 \\ 1 \end{bmatrix} - \frac{\langle u_1, v_0 \rangle}{||v_0||_2^2} v_0 = 
\begin{bmatrix} 0 \\ 1 \end{bmatrix} - \begin{bmatrix} 0 \\ 0 \end{bmatrix}$

$r_{11} = ||w_1||_2 = 1$

$r_{01} = \langle v_0, u_1 \rangle = 0$

$v_1 = \frac{w_1}{||w_0||_2} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$



Dermed bliver QR-dekomponeringen af $A$:

$$A = QR = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} \begin{bmatrix} 2 & 0 \\ 0 & 1 \end{bmatrix}$$

## (b)



```python
# Funktion som bestemmer QR dekomponering vha klassisk Gram schmidt
def klassiskGramSchmidt(A):
    # Init af matricer
    rækker, søjler = A.shape
    Q = np.ndarray((rækker, søjler))
    R = np.ndarray((rækker, rækker))
    R = np.triu(R)
    # Kører igennem hver søjle/ vektor i samlingen
    for søjle in range(søjler):
        uj = A[:, søjle] # Angiver vektoren
        wj = uj # denne + loopet trækker projektioner fra og tilføjer værdier til R
        for tidligereSøjle in range(søjle):
            vi = Q[:, tidligereSøjle]
            rij = np.vdot(uj, vi)
            R[tidligereSøjle, søjle] = rij
            wj -= rij*vi
        # Normaliserer wj til vj, og gemmer længden i R
        rjj = np.linalg.norm(uj)
        R[søjle, søjle] = rjj
        Q[:, søjle] = wj/rjj
    
    return Q, R
```

```python
A = np.array([[2, 0],
              [0, 1]], dtype = float)

klassiskGramSchmidt(A)
```

```python
Q, R = np.linalg.qr(A)
Q, R
```

```python
A = np.array([[1, 0, 1],
              [0, 1, 1], 
              [1, 1, 0]], dtype = float)

klassiskGramSchmidt(A)
```

```python
Q, R = np.linalg.qr(A)
Q, R
```

## 7.6



Simpsons tilnærmelse ser således ud for ulige $n$:



Denne vil højst have en fejl af størrelsesorden:



Hvis vi gerne vil have at vi kan garantere at de første 5 decimaler er korrekte, så skal vi have en fejl hvor følgende gælder:

$$\frac{1}{90} h^4 (b-a) \max\{|f''''(x)||a \leq x \leq b|\} < 0.00001 = 10^{-5}$$

Vi ved at:

$$a = \frac{1}{5}, b = 5$$
$$\max\{|f''''(x)||a \leq x \leq b|\} = $$


Nu skal jeg blot løse uligheden:

$$\frac{\frac{24}{5}}{90} h^4 * 4066.4 < 0.00001$$
$$\Updownarrow$$
$$\frac{24}{450} h^4 < \frac{0.00001}{4066.4}$$
$$\Updownarrow$$
$$h^4 < \frac{0.00001*450}{4066.4*24} \approx 4.610958095 * 10^{-8}$$

$$h < (4.610958095 * 10^{-8})^{1/4} = 0.014653716793271623$$

```python
def simpsonInt(f, a, b):
    # Angivelse af hvor stor intervallængden (h) må være.
    hneeded = 4.610958095*10**(-8)
    hneeded = 0.014653716793271623

    # Beregning af antal datapunkter der er behov for.
    noDataPoints = (b-a)/hneeded

    x, h = np.linspace(a, b, int(round(noDataPoints, 0)) + 1, retstep=True)

    y = f(x)

    # Sum af boxe
    totalAvgBoxHeight = 1/3*(y[0] + 2*np.sum(y[1:-1]) + 2*np.sum(y[1:-1:2]) + y[-1])

    # Beregning af areal
    SimpsBoxInt = totalAvgBoxHeight * h
    return SimpsBoxInt
```

```python
def f(x):
    return np.log(x)/(1 + x)
```

```python
print("Simpson integral af log(x)/(1+x) = ", simpsonInt(f, 1/5, 5))
```
