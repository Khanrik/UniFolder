```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r1 = -1/4
r2 = -1/5

x = np.linspace(-10, 10, 1000)
y1 = 0.5 * np.exp(r1 * x) + 0.5 * np.exp(r2 * x)

plt.plot(x, y1)
plt.grid()
plt.show()
```

```python
r1 = -1/4
r2 = -1/5

x = np.linspace(-10, 10, 1000)
y1 = 0.5 * np.exp(r1 * x) + 0.5 * np.exp(r2 * x)

plt.plot(x, y1)

r1 = -1/4 + 3j
r2 = -1/5 - 3j

x = np.linspace(-10, 10, 1000)
y1 = 0.5 * np.exp(r1 * x) + 0.5 * np.exp(r2 * x)

plt.plot(x, y1)
plt.grid()
plt.show()
```
