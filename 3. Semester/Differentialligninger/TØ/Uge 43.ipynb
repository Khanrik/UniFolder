### Problem 5
$y'' + 6y' + 9y = 0$

When guessing that the solution $y$ is of the form $y=e^{rx}$, the auxiliary equation of the problem is

$$
1 \cdot r^{2} + 6 \cdot r + 9 = 0
$$

Since the discriminant of the equation is $6^{2} - 4 \cdot 1 \cdot 9 = 0$, we know that the function will only have the single root

$$
r = \frac{-6}{2 \cdot 1} = -3
$$

The general solution for the differential equation then becomes:

$$
y(x) = c_{1} e^{-3x} + c_{2} x e^{-3x}
$$
