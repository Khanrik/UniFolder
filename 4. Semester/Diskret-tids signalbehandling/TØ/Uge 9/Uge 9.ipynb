### (a)

Given that $T = \frac{1}{F_{s}} = \frac{1}{100}$ equation 6.13 is used

\begin{align*}
X(e^{ j\omega }) &= F_{s} \sum_{k = -\infty}^{\infty} X(F_{s}j\omega - F_{s}j 2\pi k)\\
&= F_{s} \sum_{k = -\infty}^{\infty} X(j (F_{s}\omega - F_{s}2\pi k))\\
&= F_{s} \sum_{k = -\infty}^{\infty} \frac{100}{100 + (F_{s}\omega - F_{s}2\pi k)^{2}}\\
&= \sum_{k = -\infty}^{\infty} \frac{100F_{s}}{100 + (F_{s}\omega - F_{s}2\pi k)^{2}}\\
&= \sum_{k = -\infty}^{\infty} \frac{10000}{100 + (100\omega - 200\pi k)^{2}}
\end{align*}

This is the spectrum.

Using 4.12 ($\omega = 2\pi \frac{F}{F_{s}} = 2\pi \frac{F}{100}$)

\begin{align*}
X(F) &= \sum_{k = -\infty}^{\infty} \frac{100F_{s}}{100 + (F_{s} (2\pi \frac{F}{F_{s}}) - F_{s}2\pi k)^{2}}\\
&= \sum_{k = -\infty}^{\infty} \frac{100F_{s}}{100 + 4\pi^{2}(F - F_{s}k)^{2}}\\
&= \sum_{k = -\infty}^{\infty} \frac{10000}{100 + 4\pi^{2}(F - 100k)^{2}}\\
\end{align*}

Since the original spectrum given was $\frac{100}{100 + \Omega^2}$ it is known that it will decay with a rate of $\frac{1}{\Omega^2}$. The following is therefore plotted using $k = -2, \dots, 2$.

```python
F = np.linspace(-150, 150, 1000)  # Frequency in Hz
X = np.zeros_like(F, dtype=np.float64)

for k in range(-2, 2):
    X += 10000 / (100 + 4 * np.pi**2 * (F + 100 * k)**2)

plt.figure(figsize=(8, 4))
plt.plot(F, np.abs(X))
plt.xlabel('F (Hz)')
plt.ylabel(r'$|X(e^{j\omega})|$')
plt.title(r'Magnitude of $X(e^{j\omega})$')
plt.grid(True)
plt.show()
```
