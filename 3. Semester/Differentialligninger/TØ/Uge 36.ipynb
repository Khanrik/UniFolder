### Opgave 43

#### (a)
If $k$ is a constant, show that a general (one-parameter)
solution of the differential equation is given by $x(t) =
1/(C - kt)$, where $C$ is an arbitrary constant.

It can be solved as a separable differential equation as follows:

\begin{align*}
\frac{dx}{dt} &= k x^{2}\\
\frac{1}{x^{2}} dx &= k dt\\
\int \frac{1}{x^{2}} \, dx &= \int k \, dt\\
-\frac{1}{x} &= kt + C'\\
x &= -\frac{1}{kt + C'}\\
x &= \frac{1}{C - kt} \qquad (C = -C')
\end{align*}

Since it matches the given solution and C is an arbitrary constant, it is a general solution.

### (b)
Determine by inspection a solution of the initial value problem $x' = k x^{2}, x(0) = 0$.

The value cannot be inserted into the previous solution since it results in a division by zero. Instead we can give the solution $x(t) = 0$. Such a constant function should have a slope of 0 at all points. Plotting the solution into the equation gives
$$
\frac{dx}{dt} = k \cdot 0^{2}
$$
which will indeed always give a slope of 0. The solution is therefore valid.

### Problem 27

#### (a) 
Verify that if $c$ is a constant, then the function defined piecewise by
$$
y(x) =
\begin{cases}
0 \qquad \qquad \; \; \; \text{for } x \leq c, \\
(x - c)^{2} \qquad \text{for } x > c
\end{cases}
$$
satisfies the differential equation $y' = 2 \sqrt{y}$ for all $x$ (including the point $x = c$). Construct a figure illustrating the fact that the initial value problem $y' = 2 \sqrt{y}, y(0) = 0$ has infinitely many different solutions.

For the case where $x \leq c$ it is true that $y(x) = 0$. The derivative of 0 is 0, so the differential equation should lead to 0.
$$
y' = 2\sqrt{0} = 2 \cdot 0 = 0
$$
This piecewise function satisfies the differential equation for $x \leq c$.

When $x > c$ we know $y(x) = (x - c)^{2}$. Taking the derivative of this leads to
$$
\sqrt{ y } = \sqrt{ (x - c)^{2} } = x - c
$$

\begin{align*}
y'(x) &= \frac{d}{dx} (x-c)^{2}\\
&= \frac{d}{dx} (x^{2} + c^{2} - 2cx)\\
&= 2x + 0 - 2c\\
&= 2(x - c)\\
&= 2\sqrt{ y }
\end{align*}

Since this is the same as the differential equation, the piecewise function satisfies the differential equation for all $x$.

```python
n = 20
x_values = np.linspace(-3, 3, n)
y_values = np.linspace(0, 3, n)
X, Y = np.meshgrid(x_values, y_values)

DX = 1
DY = 2 * np.sqrt(Y)

N = np.sqrt(DX**2 + DY**2)*4
DX = DX / N
DY = DY / N

plt.quiver(X, Y, DX, DY, angles='xy', scale_units='xy', scale=1)
plt.xlim([-3, 3])
plt.ylim([0, 3])
plt.grid(True)
plt.show()
```

#### (b)
For what values of $b$ does the initial value problem $y' = 2\sqrt{y}, y(0) = b$ have
(i) no solution, (ii) a unique solution that is defined for all
$x$?

##### (i)

To determine this the equation first has to be solved as a seperable differential equation
\begin{align*}
\frac{dy}{dx} &= 2 \sqrt{ y }\\
\frac{1}{\sqrt{ y }} dy &= 2 dx\\
\int y^{-\frac{1}{2}} \, dy &= \int 2 \, dx\\
2 y^{\frac{1}{2}} &= 2x + C_{1}\\
\sqrt{ y } &= x + \frac{C_{1}}{2}\\
y &= (x + C_{2})^{2} \qquad C_{2} = \frac{C_{1}}{2}\\
\end{align*}

Inserting the initial value gives
\begin{align*}
(0 + C_{2})^{2} &= b\\
C_{2}^{2} = b
\end{align*}

This means that $b$ has to be positive, since it has to be equal to a square. Therefore if $b < 0$ the initial value problem has no solution.

##### (ii)

We now for solutions $b$ can't be negative, so we start looking at $b=0$
$$
(0 + C_{2})^{2} = 0
$$
In this case $C_{2} = 0$, which means that the solution would just be
$$
y = (x + 0)^{2} = x^{2}
$$
which has a unique solution defined for all values of $x$.

If $b > 0$ then $C_{2} = \pm\sqrt{b}$ which leads to the solution
$$
y = (x \pm \sqrt{b})^{2}
$$
For any given value of $b > 0$ this will also have a unique solution defined for all values of $x$.

The initial value problem will then have a unique solution defined for all value of $x$ if $b \geq 0$.

### Problem 28
Verify that if $k$ is a constant, then the function $y(x) = k x$ satisfies the differential equation $xy' = y$ for all $x$. Construct a slope field and several of these straight line solution curves. Then determine (in terms of $a$ and $b$) how many different solutions the initial value problem $xy' = y, y(a) = b$ has—one, none, or infinitely many.

To verify this, the derivative of the function is found and inserted into the differential equation whereby the equation should be satisfied
$$
\frac{d}{dx} kx = k
$$
$$
x \cdot k = k \cdot x
$$
Due to the law of commutativity this is true, which means that the function satisfies the differential equation.

```python
n = 20
x_values = np.linspace(-5, 5, n)
y_values = np.linspace(-5, 5, n)
X, Y = np.meshgrid(x_values, y_values)

DX = 1
DY = Y/X

N = np.sqrt(DX**2 + DY**2)*3
DX = DX / N
DY = DY / N

plt.quiver(X, Y, DX, DY, angles='xy', scale_units='xy', scale=1)
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.grid(True)

x_1 = x_values
y_1 = x_values * 2
plt.plot(x_1, y_1, color = 'red')

x_2 = x_values
y_2 = x_values * 4
plt.plot(x_2, y_2, color = 'green')

x_3 = x_values
y_3 = x_values * 8
plt.plot(x_3, y_3, color = 'blue')

plt.show()
```

The solution function can be substituted into $y(a)$, which allows for isolating the constant $k$ giving
\begin{align*}
y(a) &= b\\
k \cdot a &= b\\
k &= \frac{b}{a}
\end{align*}

There are now 3 cases. If $a \neq 0$, then there are infinitely many solutions of the form
$$
y(x) = \frac{b}{a} x
$$

If $a = 0$ and $b = 0$ there is one solution
$$
y(x) = 0
$$

If $a = 0$ and $b \neq 0$ there are no solution since $\frac{b}{0}$ would be undefined.
