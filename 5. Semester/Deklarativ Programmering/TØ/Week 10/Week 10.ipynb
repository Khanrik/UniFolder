## 1) Consider the following family knowledge base (this program is available in the file “family.pl”):
```pl
/* FACTS */
mother(trude, sally).
father(tom, sally).
father(tom, erica).
father(mike, tom).

/* RULES */
sibling(X, Y) :- parent(Z, X), parent(Z, Y).
parent(X, Y) :- father(X, Y).
parent(X, Y) :- mother(X, Y).
```

### (a) Use Prolog to answer the following queries:

#### i. Is Trude the mother of Sally?

```
?- mother(trude, sally).
true.
```

#### ii. Who is the father of Erica?

```
?- father(Who, erica).
Who = tom.
```

#### iii. Who is the child of Tom?

```
?- father(tom, Who).
Who = sally ;
Who = erica.
```

#### iv. Who is the sibling of Sally?

```
?- sibling(Who, sally).
Who = sally ;
Who = erica ;
Who = sally.
```

#### v. Is Mike the grandpa of Sally?

```
?- father(X, sally), father(mike, X).
X = tom.
```

#### vi. Who is a grandparent of Sally?

```
?- parent(X, sally), parent(Who, X).
X = tom,
Who = mike ;
false.
```

### (b) Notice that Sally is included as her own sibling in the query (iv) above. Fix the rules in the knowledge base so that people cannot be their own siblings (i.e. the sibling relation should be irreflexive).

An irreflexive rule is added as follows
```pl
X \= Y
```
Which makes it return
```
?- sibling(Who, sally).
Who = erica ;
false.
```

#### i. Define the grandparent relation.

```pl
grandparent(X, Y) :- parent(X, Z), parent(Z, Y).
```

#### ii. Define the ancestor relation.

```pl
ancestor(X, Y) :- parent(X, Y); parent(Z, Y), ancestor(X, Z).
```

## (3) In the lecture we discussed Constraint Logic Progrmming over finite domains, CLP(FD) Consider the following OCaml implementation of the factorial function:
```ocaml
let rec factorial n =
match n with
0 -> 1
| _ -> n * factorial (n - 1)
```

### (a) Implement this as a Prolog predicate factorial using CLP(FD) arithmetic constraints.

```pl
:- use_module(library(clpfd)).

factorial(0, 1).
factorial(N, F) :- N #> 0, N1 #= N-1, factorial(N1, F1), F #= N * F1.
```

### (b) Demonstrate your implementation by running the following queries (pressing spacebar a few times to get a handful of solutions for each query)
```
?- factorial(25, F).

?- factorial(N, 5040).

?- factorial(N, F).
```

```
?- factorial(25, F).
F = 15511210043330985984000000 .

?- factorial(N, 5040).
N = 7 .

?- factorial(N, F).
N = 0,
F = 1 .
```
