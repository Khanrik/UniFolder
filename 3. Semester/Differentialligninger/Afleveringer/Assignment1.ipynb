## Problem 1
Consider the differential equation $dy/dx = (y-3)^{2}$.

### 1.
Find the general solution of the differential equation and the singular solution that is not included in the general solution.

First the general solution is found by solving it as a seperable differential equation:

\begin{align*}
\frac{dy}{dx} &= (y-3)^{2}\\
\frac{1}{(y-3)^{2}}dy &= 1 dx\\
\int \frac{1}{(y-3)^{2}} \, dy &= \int 1 \, dx\\
- \frac{1}{y-3} &= x + C\\
y - 3 &= - \frac{1}{x+C}\\
y &= 3 - \frac{1}{x + C} &\text{(the general solution)}
\end{align*}

It can now be seen that since the right hand side of the general solution features 3 subtracted by a fraction. Since only denominator of the fraction contains variables and the numerator is not 0, the fraction will never be 0 and the solution will never reach

$$
y = 3 - 0
$$

This means that the singular solution that is not included in the general solution, which was found, is $y = 3$.

### 2.

Since $y(0) \neq 3$, the inital value problem can be solved using the general solution from before. This is done by substituting $y$ for 4 and $x$ for 0.

\begin{align*}
4 &= 3 - \frac{1}{0 + C}\\
1 &= - \frac{1}{C}\\
C &= -1
\end{align*}

The solution for the intial value is then
$$
y = 3 - \frac{1}{x - 1}
$$

```python
#Define boundaries of x and y in the slope field plot
minx = -4.
maxx = 6.
miny = -4.
maxy = 10.

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Create and normalize direction vectors
u = np.ones(x.shape)
v = (y-3)**2
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()

xs_new = np.linspace(minx, maxx, 1000)
y_sol = 3 - 1/(xs_new-1)

ax.plot(xs_new, y_sol, color='red')
```

The solution consists of two parts, which are disconnected by an asymptote at $x=1$. This disconnection is due to the fact that the denominator in fraction of the solution would become 0 at $x=1$. The solution function is therefore not defined there. When $x$ approaches $-\infty$ the solution approaches $y = 3$, but it will never be reached since it is not defined for the general solution that was used. Toward $+\infty$ the function of the solution also approaches $y=3$, but since the inital value problem was for $y(0)$, the solution only refers to the left side of asymptote and not the right side, which is a solution for a different initial value problem.

### 3.
What does Theorem 1, Existence and Uniqueness of Solutions say about the solution(s) of the initial value problem $y' = (y − 3)^{2}, y(a) = b$ for arbitrary choices of $a$ and $b$?

The theorem states that for some open interval $I$ there is one and only one solution for an initial value problem. Since we have found a solution that is not defined for x=1, the solution is defined only on the open interval $I = (-\infty, 1)$. This means that the right side of the function is not part of the solution, but that it is a solution to a different initial value problem, which is defined on the open interval $I = (1, \infty)$. This can be generalized by seeing that for any choice of $a$ and $b$, the constant $C$ can be described as

\begin{align*}
b &= 3 - \frac{1}{a + C}\\
b-3 &= - \frac{1}{a + C}\\
(a + C)(b - 3) &= -1\\
a + C &= - \frac{1}{b - 3}\\
C &= \frac{1}{3 - b} - a
\end{align*}

The denominator of the solution would then approach 0 as $x$ approaches $\frac{1}{b - 3} + a$, which means that there will be an asymptote there. Like before the same function would then be two different solution on the open intervals $I = (-\infty, \frac{1}{b - 3} + a)$ and $I = (\frac{1}{b - 3} + a, \infty)$ respectively.

There is still the special case for $y(a) = 3$, which is not defined for the general solution. Here there is only a single function, $y(x) = 3$, counting as a single solution to the problem defined for all values of $x$ in the open interval $I = (-\infty, \infty)$.

## Problem 2
Consider the differential equation $y' = (y − 2) \cos(x)$.

### 1. 
Solve the differential equation by hand for the initial value $y(0) = 0$. Account for all intermediate steps in the solution.

First the general solution is found by solving it as a seperable differential equation:

\begin{align*}
\frac{dy}{dx} &= (y − 2) \cos(x)\\
\frac{1}{y-2} dy &= \cos(x) dx\\
\int \frac{1}{y-2} \, dy &= \int \cos(x) \, dx
\end{align*}

Here substitution can be used to solve the integral on the left side, but it can also simply be solved by knowing the derivative $\frac{d}{dx} \ln(f(x)) = \frac{1}{f(x)} \cdot f'(x)$. Since the $\frac{d}{dy} y-2 = 1$, the integral of $y-2$ simply becomes $\ln(y-2)$. 

\begin{align*}
\ln(y-2) + C_{y} &= \sin(x) + C_{x}\\
\ln(y-2) &= \sin(x) + C_{1} &C_{1} = C_{x} - C_{y}\\
y-2 &= e^{\sin(x) + C_{1}}\\
y-2 &= e^{\sin(x)} \cdot e^{C_{1}}\\
y-2 &= C_{2} e^{\sin(x)} &C_{2} = e^{C_{1}}\\
y &= C_{2} e^{\sin(x)} + 2 &\text{(the general solution)}
\end{align*}

The initial value can then be inserted into the general solution to find $C_{2}$

\begin{align*}
0 &= C_{2} e^{\sin(0)} + 2\\
-2 &= C_{2} e^{0}\\
C_{2} &= -2
\end{align*}

Inserting the value found for $C_{2}$ back into the general solution gives the particular solution for the initial value problem

$$
y = -2 e^{\sin(x)} + 2 \qquad \text{(the particular solution)}
$$



### 2. 
Discuss the solutions to the differential equation if other initial values than 0 are used, i.e., the $y(0) = b, b \neq 0$ case. Include plots of the slope field and representative solutions.

As can be seen when finding $C_{2}$ in the previous part, it only depends on the value of $b$ since $e^{\sin(x)} = 1$. For any value $b$ the constant $C_{2}$ can be found as
$$
C_{2} = b - 2
$$
making it very easy to find solutions using for loops in python.

This also means that the solution for the initial value $y(0) = 2$, would become the constant function $y = (2-2)e^{\sin(x)} + 2 = 2$. Every other function however will grow and shrink exponentially with a periodicity defined by the sinus function in the exponent. Various solutions (excluding $y(0)=0$) can be seen below.

```python
#Define ODE
def df(x,y):
    return (y-2) * np.cos(x)

def sol(x,c):
    return c * np.exp(np.sin(x)) + 2

#Show scipy numerical solution True/False
show_solution = True

#Define boundaries of x and y in the slope field plot
minx = -10.
maxx = 10.
miny = -12.5
maxy = 12.5

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Define the initial value(s) - Even if only one: Make a 1D array
x0 = minx
y0_values = np.array([-2, -1, 1, 2, 3, 4, 5])

#Create and normalize direction vectors
u = np.ones(x.shape)
v = df(x,y)
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
if show_solution:
    for y0 in y0_values:
        #Solution
        xsol = np.linspace(x0, maxx, xsteps*100)
        c = y0-2
        ysol = sol(xsol, c)
        ax.plot(xsol, ysol, label=f'$y(0)$ = {y0}')
        ax.legend(loc='upper left', fontsize=15)
ax.set_title("Slope Field")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()
# plt.savefig("slope_field")
plt.show()
```

## Problem 3
Consider the differential equation $y' = 2x − y$.

### 1.
Solve the differential equation with the initial value $y(0) = 3$.

First the general solution is found. This is done by getting the equation on the form $dy/dx + P(x)y = Q(x)$:

\begin{align*}
\frac{dy}{dx} &= 2x - y\\
\frac{dy}{dx} + y &= 2x\\
\end{align*}

The general solution can now be found using Theorem 1.5.1:

\begin{align*}
y(x) &= e^{-\int 1 \, dx } \left[ \int \left( 2x e^{\int 1 \, dx } \right) \, dx + C \right]\\
&= e^{-x} \left[ \int 2xe^{x} \, dx + C \right]\\
&= e^{-x} \left[ 2x \cdot e^{x} - \int 2e^{x} \, dx + C \right] &\text{(integration by parts)}\\
&= e^{-x} [2x \cdot e^{x} - 2e^{x} + C]\\
&= 2x - 2 + Ce^{-x}\\
&= Ce^{-x} + 2x - 2 &\text{(the general solution)}
\end{align*}

The constant $C$ corresponding to the problem can be found by inserting the inital value:

\begin{align*}
3 &= C e^{-0} + 2 \cdot 0 - 2\\
3 &= C - 2\\
C &= 5
\end{align*}

Inserting the value found for $C$ back into the general solution gives the particular solution for the initial value problem

$$
y = 5e^{-x} + 2x - 2 \qquad \text{(the particular solution)}
$$

## Problem 4
Consider the logistic differential equation $y' = 2y(5 − y)$.

### 1.
Compute and plot the solutions for $y(0) = 1$ and $y(1) = 1$. What features do the two solutions have in common?

The general solution is first calculated as a seperable differential equation:

\begin{align*}
\frac{dy}{dx} &= 2y(5 - y)\\
\int \frac{1}{2y(5 - y)} \, dy &= \int 1 \, dx\\
\frac{1}{2} \int \frac{1}{y(5 - y)} \, dy &= x + C_{1}
\end{align*}

The left side can be rewritten using partial fractions as follows

\begin{align*}
\frac{A}{y} + \frac{B}{5-y} &= \frac{1}{y(5-y)}\\
\frac{A \cdot y(5 - y)}{y} + \frac{B \cdot y(5 - y)}{5 - y} &= 1\\
A \cdot (5 - y) + B \cdot y &= 1\\
A \cdot 5 + B \cdot 0 &= 1 \qquad y=0 \text{ is chosen to find A since any y value is valid}\\
A &= \frac{1}{5}\\
A \cdot 0 + B \cdot 5 &= 1 \qquad y=5 \text{ is chosen to find B}\\
B &= \frac{1}{5}
\end{align*}

Inserting the partial fractions back in the caluculation yields

\begin{align*}
\frac{1}{2} \int \frac{1}{5y} + \frac{1}{5 (5 - y)} \, dy &= x + C_{1}\\
\frac{1}{10} \int \frac{1}{y} + \frac{1}{5 - y} \, dy &= x + C_{1}\\
\frac{1}{10} \cdot (\ln(y) - \ln(5 - y)) &= x + C_{1}\\
\ln(y) - \ln(5 - y) &= 10x + 10C_{1}\\
e^{\ln(y) - \ln(5 - y)} &= e^{10x + 10C_{1}}\\
\frac{y}{5-y} &= e^{10x} \cdot e^{10C_{1}}\\
y &= e^{10x} \cdot e^{10C_{1}} \cdot (5 - y) &C_{2} = e^{10C_{1}}\\
y &= e^{10x} \cdot C_{2} \cdot 5 - e^{10x} \cdot C_{2} \cdot y\\
y + e^{10x} \cdot C_{2} \cdot y &= 5 \cdot e^{10x} \cdot C_{2}\\
y \cdot (1 + e^{10x} \cdot C_{2}) &= 5 \cdot e^{10x} \cdot C_{2}\\
y &= \frac{5 \cdot e^{10x} \cdot C_{2}}{1 + e^{10x} \cdot C_{2}}\\
y &= \frac{5 \cdot e^{10x} \cdot \frac{1}{C_{3}}}{1 + e^{10x} \cdot \frac{1}{C_{3}}} &C_{3} = \frac{1}{C_{2}}\\
y &= \frac{5 \cdot e^{10x}}{C_{3} + e^{10x}} &\text{(the general solution)}
\end{align*}

```python
#Define boundaries of x and y in the slope field plot
minx = -10.
maxx = 10.
miny = -10.
maxy = 10.

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Create and normalize direction vectors
u = np.ones(x.shape)
v = (y-3)**2
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))

xs_new = np.linspace(minx, maxx, 1000)
y_sol = (np.exp(10*xs_new)*5*0.25)/(1+np.exp(10*xs_new)*0.25)
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.plot(xs_new, y_sol, color='red')
```
