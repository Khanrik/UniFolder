```python
def klassisk_gram_schmidt(a):
    _, k = a.shape
    q = np.empty_like(a)
    r = np.zeros((k, k))
    for j in range(k):
        r[:j, [j]] = q[:, :j].T @ a[:, [j]]
        w = a[:, [j]] - q[:, :j] @ r[:j, [j]]
        r[j, j] = np.linalg.norm(w)
        q[:, [j]] = w / r[j, j]
    return q, r
```

```python
def forbedret_gram_schmidt(a):
    _, k = a.shape
    q = np.copy(a)
    r = np.zeros((k, k))
    for i in range(k):
        r[i, i] = np.linalg.norm(q[:, i])
        q[:, i] /= r[i, i]
        r[[i], i+1:] = q[:, [i]].T @ q[:, i+1:]
        q[:, i+1:] -= q[:, [i]] @ r[[i], i+1:]
    return q, r
```

```python
def house(x):
    u = x / np.linalg.norm(x)
    eps = +1 if u[0, 0] >= 0 else -1
    s = 1 + np.abs(u[0, 0])
    v = (eps / s) * u
    v[0, 0] = 1
    return v, s

def householder_qr_data(a):
    data = np.copy(a)
    _, k = a.shape
    s = np.empty(k)
    for j in range(k):
        v, s[j] = house(data[j:, [j]])
        data[j:, j:] -= (s[j] * v) @ (v.T @ data[j:, j:])
        data[j+1:, [j]] = v[1:]
    return data, s

def householder_qr(a):
    data, s = householder_qr_data(a)
    n, k = a.shape
    r = np.triu(data[:k, :k])
    q = np.eye(n, k)
    for j in reversed(range(k)):
        x = data[j+1:, [j]]
        v = np.vstack([[1], x])
        q[j:, j:] -= (s[j] * v) @ (v.T @ q[j:, j:])
    return q, r
```

```python
rng = np.random.default_rng()
n = 100
u, _, vt = np.linalg.svd(rng.random((n, n)))
i = np.arange(n)
s = np.array(2.0 ** (-i))
a = u @ np.diag(s) @ vt
print(s[:5])
```

```python
qk, rk = klassisk_gram_schmidt(a)
qf, rf = forbedret_gram_schmidt(a)
qh, rh = householder_qr(a)
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.plot(i, rk[i,i], 'o', label='klassisk Gram-Schmidt')
ax.plot(i, rf[i,i], 'x', label='forbedret Gram-Schmidt')
ax.plot(i, rh[i,i], '.', label='Householder_QR')
ax.plot(i, s[i], '.', label='Singulærværider')
ax.set_xlabel('i')
ax.set_ylabel('r[i,i]')
ax.legend()
```
