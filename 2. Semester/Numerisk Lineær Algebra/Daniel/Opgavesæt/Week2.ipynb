## a)

```python
v = np.array([0.0, 1.0], dtype = float).T
```

Magnitude kan findes på forskellige måder:

$$|v| = \sqrt{x^2 + y^2} = \sqrt{\langle v, v\rangle}$$

Eller i numpy ved at bruge, np.linalg.norm(v).

```python
length = np.sqrt(np.vdot(v, v))

print("|v| =", length)

angle = np.arctan2(v[1], v[0])

print("Angle in radians:", angle, "rad;", "In radians as multiple of pi:", angle/np.pi, "pi;", "In degrees:", angle/(2*np.pi)*360, "degrees")
```

# Opg 2.3



Som kan ses, så laves der færre operationer i udregningen i $z$, da vi blot indsætter nogle x-værdier. Da der laves færre operationer, så sker der nok den mindste fejl i denne beregning.

```python
x = np.linspace(0.0, 2.0, 1000)
y = x**8 - 8*x**7 + 28*x**6 - 56*x**5 + 70*x**4 - 56*x**3 + 28*x**2 - 8*x + 1
z = (x - 1)**8

# Fejlen af y ift z
yfejl = y - z

# Relativ fejl af y ift z

yrelfejl = (y - z) / z

fig, ax = plt.subplots()
ax.plot(x, yfejl, label = "Absolut fejl")

fig, bx = plt.subplots()
bx.plot(x, yrelfejl, label = "Relativ fejl")

ax.legend()
bx.legend()
```

$$e_0 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$$

$$e_0 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$$

Jeg opstiller to ligninger hvor dette gælder:

$$Ae_0 = \begin{bmatrix} a \\ c \end{bmatrix} = \begin{bmatrix} 4 \\ -3 \end{bmatrix}$$

$$Ae_1 = \begin{bmatrix} b \\ d \end{bmatrix} = \begin{bmatrix} 2 \\ 5 \end{bmatrix}$$

Altså må det gælde at $A = \begin{bmatrix} 4 & 2 \\ -3 & 5 \end{bmatrix}$

## b)



Hvis $A$ er en spejlings matrix står den på formen:

$$A = \begin{bmatrix} c & s \\ s & -c \end{bmatrix}, c^2 + s^2 = 1$$

Jeg opstiller ligningen:

$$A \begin{bmatrix} 4 \\ -3 \end{bmatrix} = \begin{bmatrix} 4c - 3s \\ 4s + 3c \end{bmatrix} = \begin{bmatrix} 5 \\ 0 \end{bmatrix}$$

Vi har nu to ligninger med to ubekendte:

$$4c - 3s = 5 \Leftrightarrow c = \frac{5 + 3s}{4}$$
$$4s + 3c = 0 \Leftrightarrow 4s + 3\frac{5 + 3s}{4} = \frac{16s + 15 + 9s}{4} = \frac{25s + 15}{4} = 0 \Leftrightarrow s = -\frac{3}{5}$$
$$c = \frac{5-3\frac{3}{5}}{4} = \frac{16}{20} = \frac{4}{5}$$

Dermed er:

$$A = \begin{bmatrix} \frac{4}{5} & -\frac{3}{5} \\ -\frac{3}{5} & -\frac{4}{5} \end{bmatrix}$$
