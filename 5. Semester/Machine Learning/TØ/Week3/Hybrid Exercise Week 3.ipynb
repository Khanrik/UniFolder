## 1) Estimate the mean of the Gaussian distribution, μM L, from the available data points by using the maximum likelihood (ML) method.

```python
# Lecture 5 slide 9 (slide 131)
# Equation 1.55 in book
xs = np.array([0.0557, 1.8529, 1.9311, 2.1905, 3.8884, 4.4384])
N = len(xs)
mu_ML = 1/N * sum(xs)
print("Using MLE, the mean of the gaussian distribution is calculated as", mu_ML)
```

## 2) Estimate the variance of the Gaussian distribution. σ2 M L, from the available data points by using the maximum likelihood (ML) method.

```python
# Equation 1.56 in book
sigma2_ML = 1/N * sum((xs - mu_ML)**2)
print("Using MLE, the variance of the gaussian distribution is calculated as", sigma2_ML)
```

## 3) Comment on whether the above estimates of the mean μM L and the variance σ2 M L are biased or unbiased?

The maximum likehood function will underestimate the variance of the distribution by a factor of $\frac{N-1}{N}$ (equation 1.58), meaning that it is biased (since it is based on the sample mean and not population mean). The mean is however an unbiased estimate of the mean of the distribution since it is averaged out over all data points.

## 4) If one or both the above estimates obtained in the previous step are biased, how would you make those unbiased?

Only the variance was biased, and it can be made unbiased using equation 1.59 in the book:
$$
\tilde{\sigma}^2_{ML} = \frac{N}{N-1}\sigma^2_{ML} = \frac{1}{N-1}\sum_{n=1}^{N}(x_n - \mu_{ML})^2
$$

## 5) Now, assume that the actual variance of the Gaussian distribution that led to the data points in Fig. 1 is known and is σ2 = 1. Using that information and the same data points as in Fig. 1, estimate the mean of the Gaussian distribution μBayes through the Bayesian approach by using the following conjugate form for the prior distribution of mean: p(μ) = N (μ|μ0 = 4, σ2 0 = 3).

```python
# Lecture 6 slide 15 (Slide 164)
sigma2 = 1
sigma2_0 = 3
mu_0 = 4
mu_N = sigma2 / (N * sigma2_0 + sigma2) * mu_0 + (N * sigma2_0) / (N * sigma2_0 + sigma2) * mu_ML
print("Using Bayesian estimation, the mean of the gaussian distribution is calculated as", mu_N)
```

## 6) Given that the information about the prior density p(μ) is reasonable, which of the above density estimation approaches (i.e., ML or Bayesian) do you expect will perform well in the following settings: a) for smaller values of N (e.g., for N = 6 as in Fig. 1); b) for N → ∞.

For $N = 6$, the Bayesian approach is expected to perform better.

For $N \to \infty$, both approaches are expected to perform equally well, but MLE is simpler to compute.

## 7) Briefly mention the reasons behind your answers in the previous question (6).

In a small dataset such as $N = 6$, MLE can be easily influenced by outliers, and therefore is likely to give biased and inaccurate estimates. The Bayesian approach incorporates prior knowledge, and assuming that it is reasonable, it should perform a lot better.

When $N \to \infty$, the influence of outliers is diminished and both estimations should converge to the same values. MLE is simpler to compute, and therefore preferred in this case.
