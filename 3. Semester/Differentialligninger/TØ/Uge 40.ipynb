### Problems 27
$y' = x^{2} + y^{2} - 1, \, y(0) = 0; \, y(2) = ?$

```python
def df(x, y):
    return x**2 + y**2 - 1

def euler(func, x0, y0, h, x_end):
    Y = [y0]
    for i in range(int((x_end - x0) / h)+1):
        x_n = x0 + i * h
        Y.append(Y[i] + h * func(x_n, Y[i]))
    return Y[-1]

x0 = y0 = 0
x_end = 2
h = 0.1

for i in range(0, 10):
    new_h = h/2**i
    print(f"With step size h = {new_h} {' ' * (12 - len(str(round(new_h, 10))))} the value y({x_end}) = {euler(df, x0, y0, new_h, x_end)}")
```

### Problems 29
Consider the initial value problem
$$
7x \frac{dy}{dx} + y = 0, \qquad y(-1) = 1
$$

#### (a)
Solve this problem for the exact solution
$$
y(x) = - \frac{1}{x^{1/7}}
$$
which has an infinite discontinuity at $x = 0$.

Find general solution as separable equations

\begin{align*}
7x \frac{dy}{dx} + y &= 0\\
7x \frac{dy}{dx} &= -y\\
7x dy &= -y dx\\
- \frac{1}{y} dy &= \frac{1}{7x} dx\\
- \int \frac{1}{y} \, dy &= \int \frac{1}{7x} \, dx\\
-\ln(y) &= \frac{1}{7}\ln(7x) + c_{1}\\
y &= e^{-\frac{1}{7}\ln(7x) - c_{1}}\\
y &= (e^{\ln(x)})^{- 1/7} \cdot e^{-c_{1}}\\
y &= x^{- 1/7} \cdot c_{2} \qquad c_{2} = e^{-c_{1}}\\
y &= \frac{c_{2}}{x^{1/7}}
\end{align*}

Insert inital value to find exact solution

\begin{align*}
1 &= \frac{c_{2}}{(-1)^{1/7}}\\
c_{2} &= -1
\end{align*}

The exact solution for the initial value problem is then $y(x) = -\frac{1}{x^{1/7}}$, which indeed has a discontinuity at $x = 0$, since the denominator becomes 0.

## Section 2.5

Use the improved Euler method with a computer system to find the desired solution values in Problems 27 and 28. Start with step size $h = 0.1$, and then use successively smaller step sizes until successive approximate solution values at $x = 2$ agree rounded off to four decimal places.

### 27.
$y' = x^{2} + y^{2} - 1, \, y(0) = 0; \, y(2) = ?$

```python
def df(x, y):
    return x**2 + y**2 - 1

def improved_euler(func, x0, y0, h, x_end):
    Y = [y0]
    for i in range(int((x_end - x0) / h)+1):
        x_n = x0 + i * h
        k1 = func(x_n, Y[i])
        u_next = Y[i] + h * k1
        k2 = func(x_n + h, u_next)
        Y.append(Y[i] + h * (1/2) * (k1 + k2))
    return Y[-1]

x0 = y0 = 0
x_end = 2
h = 0.1

for i in range(0, 15):
    print(f"With step size h = {round(h/2**i, 10)} {' ' * (10-i if i < 10 else 2)} the value y({x_end}) = {improved_euler(df, x0, y0, h/2**i, x_end)}")
```

## Section 2.6

Use the Runge–Kutta method with a computer system to find the desired solution values in Problems 27 and 28. Start with step size $h = 1$, and then use successively smaller step sizes until successive approximate solution values at $x = 2$ agree rounded off to five decimal places.

### 27.
$y' = x^{2} + y^{2} - 1, \, y(0) = 0; \, y(2) = ?$

```python
def df(x, y):
    return x**2 + y**2 - 1

def runge_kutta(func, x0, y0, h, x_end):
    Y = [y0]
    for i in range(int((x_end - x0) / h)+1):
        x_n = x0 + i * h
        k1 = func(x_n, Y[i])
        k2 = func(x_n + 1/2 * h, Y[i] + 1/2 * h * k1)
        k3 = func(x_n + 1/2 * h, Y[i] + 1/2 * h * k2)
        k4 = func(x_n + h, Y[i] + h * k3)
        Y.append(Y[i] + (h/6) * (k1 + k2 + k3 + k4))
    return Y[-1]

x0 = y0 = 0
x_end = 2
h = 1

for i in range(0, 15):
    new_h = h/2**i
    print(f"With step size h = {round(new_h, 10)} {' ' * (12 - len(str(round(new_h, 10))))} the value y({x_end}) = {runge_kutta(df, x0, y0, new_h, x_end)}")
```
