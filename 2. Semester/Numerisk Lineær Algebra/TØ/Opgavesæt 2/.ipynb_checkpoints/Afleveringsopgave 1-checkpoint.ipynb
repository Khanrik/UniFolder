En robotarm består af 4 stænger $OA$, $AB$, $BC$ og $CP$ samlet i planen således at armen kan bøjes ved ledderne $O$, $A$, $B$ og $C$. Lad $a = \overrightarrow{OA}$, $b = \overrightarrow{AB}$, $c = \overrightarrow{BC}$ og $d = \overrightarrow{CP}$. Robotarmens stilling er bestemt af matricen
$$
S = [a | b | c | d]
$$

## (a)

Vælg nogle passende værdier for indgangene i **a**, **b**, **c** og **d**, og brug derefter *matplotlib* til at lave en tegningen af robotarmen, som minder om den ovenfor.

```python
O = np.array([0, 0])
A = np.array([2, 6])
B = np.array([3.5, 6.5])
C = np.array([4, 5.5])
P = np.array([3.75, 5])

a = A - O
b = B - A
c = C - B
d = P - C

S = np.array([O, 
              a, 
              a + b, 
              a + b + c, 
              a + b + c + d])
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(S[:, 0], S[:, 1], marker='o')
```

## (b)

Bestem vektoren $\overrightarrow{OP}$ ud fra **a**, **b**, **c**, **d**.

```python
a = A - O
b = B - A
c = C - B
d = P - C

OP = a + b + c + d
OP
```

## (c)

```python
theta = 2
co = np.cos(theta)
si = np.sin(theta)

R = np.array([
    [co, -si],
    [si, co]
])

d_rot_C = R @ d

S_rot_C = np.array([O, 
                    a, 
                    a + b, 
                    a + b + c, 
                    a + b + c + d_rot_C])
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(S_rot_C[:, 0], S_rot_C[:, 1], marker='o')
```

## (d)

```python
theta = -np.pi/2
co = np.cos(theta)
si = np.sin(theta)

R = np.array([
    [co, -si],
    [si, co]
])

S_rot_A = S @ R

b_rot_A = S_rot_A[2] - S_rot_A[1]
c_rot_A = S_rot_A[3] - S_rot_A[2]
d_rot_A = S_rot_A[4] - S_rot_A[3]

S_rot_A = np.array([O, 
                    a, 
                    a + b_rot_A, 
                    a + b_rot_A + c_rot_A, 
                    a + b_rot_A + c_rot_A + d_rot_A])

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(S_rot_A[:, 0], S_rot_A[:, 1], marker='o')
```

```python
# Definer rotationsmatrix
theta = -np.pi
co = np.cos(theta)
si = np.sin(theta)

R = np.array([
    [co, -si],
    [si, co]
])

# Roter original arm position
S_rot_A = S @ R

# Få nye vectorer for b, c og d (fastholder a)
b_rot_A = S_rot_A[2] - S_rot_A[1]
c_rot_A = S_rot_A[3] - S_rot_A[2]
d_rot_A = S_rot_A[4] - S_rot_A[3]

# Definer nye punkter for robotarm
S_rot_A = np.array([O, 
                    a, 
                    a + b_rot_A, 
                    a + b_rot_A + c_rot_A, 
                    a + b_rot_A + c_rot_A + d_rot_A])

# Plot
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(S_rot_A[:, 0], S_rot_A[:, 1], marker='o')
```
