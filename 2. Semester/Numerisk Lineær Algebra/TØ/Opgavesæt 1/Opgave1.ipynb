## Opgave 1.2

Ifølge google bor du 3,1 km fra universitet og din fitness app fortæller
at du går 5,8 km/h, som er ikke så langsomt. Hvor mange minutter tager det
dig til at komme hen til universitet? Giv maksimums- og minimumsværdier til
én decimal

````python
afstand = 3.1   # km
hastighed = 5.8 # km/h
error = 0.5
````

````python
avg_tid = afstand / hastighed * 60  # min

min_tid = avg_tid - error
print(f"Minimumsværdien er: {min_tid}")
max_tid = avg_tid + error
print(f"Maksimumsværdien er: {max_tid}")
````

## Opgave 1.3

I et simpelt elektrisk kredsløb har du et AAA batteri, som yder en
spænding omkring 𝑉 =1,5 V, og en resistor med modstand 𝑅 =20 Ω±5 %. Den
elektriske strøm 𝐼 igennem 𝑅 er givet ved Ohms lov 𝑉 =𝑅 ·𝐼 og den elektriske
effekt 𝑃 er 𝑃 =𝑉 ·𝐼. Vælg en realistisk størrelse for usikkerheden i værdien 𝑉,
og bestem derefter de mulige ekstreme værdier for 𝐼 og 𝑃.

````python
volt = 1.5          # V
volt_error = 0.05 * volt
ohm = 20            # Ω
ohm_error = 0.05 * ohm
````

````python
min_amp = (volt - volt_error) / (ohm + ohm_error)           # A
print(f"Minimumsværdien for strømmen (I) er: {min_amp}")
max_amp = (volt + volt_error) / (ohm - ohm_error)           # A
print(f"Maksimumsværdien for strømmen (I) er: {max_amp}")

print() # newline

min_watt = (volt - volt_error) * min_amp                    # W
print(f"Minimumsværdien for effekten (P) er: {min_watt}")
max_watt = (volt + volt_error) * max_amp                    # W
print(f"Maksimumsværdien for effekten (P) er: {max_watt}")
````

## Opgave 1.5

Softmax kan have risiko for overflow, når eksponenten i tælleren er for stor, hvilket gør at computeren essentielt ignorerer det andet led i nævneren. For at undgå dette kan der trækkes en konstant, K, fra eksponenten. Denne konstant kan bestemmes til forskellige værdier alt efter hvor store tal der arbejdes med. Den endelige funktion bliver da:

$$
softmax(x,y) = \left( \frac{e^{x - K}}{e^{x - K} + e^{y - K}} , \frac{e^{y - K}}{e^{x - K} + e^{y - K}} \right)
$$

Som bonus kan det siges, at den $i$'te koordinat af en softmax funktion med n parametre, vil have værdien:

$$
softmax(x_{0}, x_{1}, \dots, x_{n-1}) = \left( \frac{e^{x_{0}}}{\sum_{i=0}^{n-1} e^{x_{i}}} , \frac{e^{x_{1}}}{\sum_{i=0}^{n-1} e^{x_{i}}} , \dots , \frac{e^{x_{n - 1}}}{\sum_{i=0}^{n-1} e^{x_{i}}} \right)
$$

````python
data = pd.read_csv("fish.csv")
fish = data.to_numpy()
fish[:,1] = -fish[:,1]

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(fish[:,0] , fish[:,1])
````

````python
A = np.array([
    [2.0, -1.0],
    [-1.0, 2.0]
])

tilt_fish = fish @ A

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(tilt_fish[:,0] , tilt_fish[:,1])
````

#### (i)

````python
c = np.cos(2)
s = np.sin(2)

R = np.array([
    [c, -s],
    [s, c]
])

rot_fish = fish @ R         # rotaiton 1
rot_fish = rot_fish @ R     # rotation 2

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(rot_fish[:,0] , rot_fish[:,1])
````

````python
c = np.cos(5)
s = np.sin(5)

M = np.array([
    [c, s],
    [s, -c]
])

mir_fish = fish @ M         # mirror 1
mir_fish = mir_fish @ M     # mirror 2

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot(mir_fish[:,0] , mir_fish[:,1])
````
