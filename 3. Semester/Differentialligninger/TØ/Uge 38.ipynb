## Section 1.4

Find general solutions (implicit if necessary, explicit if convenient) of the differential equations in Problems 1 through 18. Primes denote derivatives with respect to $x$.

### Problem 1
$\frac{dy}{dx} + 2xy = 0$

Standard form defines $P(x) = 2x$ and $Q(x) = 0$
\begin{align*}
v(x) &= e^{\int 2x \, dx}\\
&= e^{x^{2}}\\
\end{align*}

\begin{align*}
y &= \frac{1}{e^{x^{2}}} \int e^{x^{2}} \cdot 0 \, dx\\
&= e^{-x^{2}} \cdot C\\
&= C e^{-x^{2}}
\end{align*}


### Problem 3
$\frac{dy}{dx} = y \cdot \sin(x)$

Rewrite as 
$$
\frac{dy}{dx} - \sin(x) \cdot y = 0
$$

Standard form defines $P(x) = -\sin(x)$ and $Q(x) = 0$
\begin{align*}
v(x) &= e^{\int -\sin(x) \, dx}\\
&= e^{\cos(x)}\\
\end{align*}

\begin{align*}
y &= \frac{1}{e^{\cos(x)}} \int e^{\cos(x)} \cdot 0 \, dx\\
&= e^{-\cos(x)} \cdot C\\
&= C e^{-\cos(x)}
\end{align*}


### Problem 5
$2 \sqrt{x} \frac{dy}{dx} = \sqrt{1 - y^{2}}$

Can be solved as separable differential equation as follows
$$
\frac{1}{\sqrt{1 - y^{2}}} dy = \frac{1}{2 \sqrt{x}} dx
$$
