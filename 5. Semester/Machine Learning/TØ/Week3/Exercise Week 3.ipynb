## 1) Estimate the mean of the Gaussian distribution, μML, from the available data points by using the maximum likelihood (ML) method. (Score: 15%)

\begin{aligned}
\mu_{ML}&= \frac{1}{N}\sum^{N}_{i=1}x_{i}\\
&= \frac{1}{6}\cdot(0.0557+1.8529+1.9311+2.1905+3.8884+4.4384)\\
&= 2.3928333
\end{aligned}

## 2) Estimate the variance of the Gaussian distribution. $\sigma^{2}{ML}$, from the available data points by using the maximum likelihood (ML) method. (Score: 15%)

\begin{aligned}
\sigma_{ML}^{2}&= \frac{1}{N}\sum^{N}_{i=1}(x_{i}-\mu_{ML})^{2}\\
&= \frac{1}{6}\cdot((0.0557-2.3928333)^{2} +(1.8529-2.3928333)^{2}+(1.9311-2.3928333)^{2}+(2.1905-2.3928333)^{2}+(3.8884-2.3928333)^{2}+(4.4384-2.3928333)^{2})\\
&= 2.07148655
\end{aligned}

## 3) Comment on whether the above estimates of the mean $\mu_{ML}$ and the variance $\sigma_{ML}^{2}$ are biased or unbiased? (Score: 10%)

$\mu_{ML}$ is unbaised, while $\sigma_{ML}^{2}$ is biased downward.

## 4) If one or both the above estimates obtained in the previous step are biased, how would you make those unbiased? (Score: 10%)

Bessel's correction can be used to make the variance estimate unbiased. Instead of dividing by N, we divide by N-1. It then becomes

\begin{aligned}
\sigma_{unbiased}^{2} &= \frac{1}{N-1}\sum^{N}_{i=1}(x_{i}-\mu_{ML})^{2}\\
&= \frac{1}{5}\cdot((0.0557-2.3928333)^{2} +(1.8529-2.3928333)^{2}+(1.9311-2.3928333)^{2}+(2.1905-2.3928333)^{2}+(3.8884-2.3928333)^{2}+(4.4384-2.3928333)^{2})\\
&= 2.48578386
\end{aligned}

Using the bayesian approach the mean is given as

## 6) Given that the information about the prior density p(μ) is reasonable, which of the above density estimation approaches (i.e., ML or Bayesian) do you expect will perform well in the following settings: a) for smaller values of $N$ (e.g., for $N = 6$ as in Fig. 1); b) for $N \to \infty$. (Score: 10%)

For $N = 6$, the Bayesian approach is expected to perform better.

For $N \to \infty$, both approaches are expected to perform equally well, but MLE is simpler to compute.

## 7) Briefly mention the reasons behind your answers in the previous question (6). (Score: 15%)

In a small dataset such as $N = 6$, MLE can be easily influenced by outliers, and therefore is likely to give biased and inaccurate estimates. The Bayesian approach incorporates prior knowledge, and assuming that it is reasonable, it should perform a lot better.

When $N \to \infty$, the influence of outliers is diminished and both estimations should converge to the same values. MLE is simpler to compute, and therefore preferred in this case.
