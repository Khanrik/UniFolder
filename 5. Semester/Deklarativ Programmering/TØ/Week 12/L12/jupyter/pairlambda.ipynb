```ocaml
#use_output "dune top"
#directory "lclib"
open LamTerm
#install_printer lprinter
```

```ocaml
lparser "f @ g";;
lparser "f @ g @ h";;
lparser "f @ (g @ h)";;
lparser "x, y";;
lparser "x, y, z";;
lparser "(x, y), z"
```

```ocaml
eval_lazy "= @ (fst @ x, y) @ x";;
eval_lazy "= @ (fst @ x, y) @ y";;
eval_lazy "= @ (fst @ x, y, z, nil) @ x";;
eval_lazy "= @ (snd @ x, y) @ y";;
eval_lazy "= @ (snd @ x, y) @ x";;
eval_lazy "= @ (snd @ x, y, z, nil) @ y, z, nil"
```

```ocaml
let y = "(fun g. (fun y. g @ (y @ y)) @ (fun z. g @ (z @ z)))";;
let conc = "(fun f xs ys. ? @ (= @ xs @ nil) @ ys @ (fst @ xs), (f @ (snd @ xs) @ ys))";;
let yconc = y ^ " @ " ^ conc;;
lparser yconc
```

```ocaml
eval_lazy (yconc ^ " @ 1, 2, nil @ 3, 4, nil")
```

```ocaml
let foldr = "(fun f g c xs. ? @ (= @ xs @ nil) @ c @ (g @ (fst @ xs) @ (f @ g @ c @ (snd @ xs))))";;
let yfoldr = y ^ " @ " ^ foldr;;
lparser yfoldr
```

```ocaml
eval_lazy (yfoldr ^ " @ (fun x y. + @ x @ y) @ 0 @ 1, 2, 3, nil");;
yfoldr ^ " @ (fun x y. + @ x @ y) @ 0 @ 1, 2, 3, nil"
```
