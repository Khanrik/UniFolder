### (b) Ved brug af en tynd 𝑄𝑅-dekomponering konstrueret via den forbedrede Gram-Schmidt proces, bestem i python den mindste kvadraters løsning til systemet $𝐴𝑐 = 𝑏$. {-}

````python
# Fra afsnit 15.3 i notesættet
def forbedret_gram_schmidt(a):
    _, k = a.shape
    q = np.copy(a)
    r = np.zeros((k, k))
    for i in range(k):
        r[i, i] = np.linalg.norm(q[:, i])
        q[:, i] /= r[i, i]
        r[[i], i+1:] = q[:, [i]].T @ q[:, i+1:]
        q[:, i+1:] -= q[:, [i]] @ r[[i], i+1:]
    return q, r

# Fra afsnit 16.4 i notesættet
def back_subs(r, c):
    _, n = r.shape
    x = np.empty((n, 1))
    for i in reversed(range(n)):
        x[i] = (c[i] - r[[i], i+1:] @ x[i+1:]) / r[i, i]
    return x
````

````python
# Resultaterne fra QR-dekomponeringen samt b indsættes i ligning 16.4 fra notesættet
q, r = forbedret_gram_schmidt(A)
QTb = q.T @ b
c = back_subs(r, QTb)
print(f"Den mindste kvadraters løsning bliver da: \n{c}")
````

### (d) Brug en potensmetode til at bestemme egenværdien af $𝐴$, som er størst numerisk, og en tilhørende egenvektor. {-}

````python
# Fra eksempel 24.1 i notesættet
def potens_skridt(a, w):
    v = a @ w
    v /= np.linalg.norm(v)
    lambda_ny = np.vdot(v, a @ v)
    return lambda_ny, v
````

````python
# Følger eksempel 24.1 i notesættet

rng = np.random.default_rng()

m, _ = A.shape

v3 = rng.standard_normal((m, 1))
n = 50
lambda_out = np.zeros(n)

for k in range(n):
    lambda_out[k], v3 = potens_skridt(A, v3)

lambda3 = lambda_out[-1]

print("Den største numeriske egenværdi er", lambda3)
print("\nDens tilhørende egenvektor er\n", v3)
````

````python
# Opstiller fundne værdier
v0 = np.array([1, 1, 1, 1])[:, np.newaxis]
v1 = np.array([1, -1, 1, -1])[:, np.newaxis]
v2 = np.array([-1, 1, 1, -1])[:, np.newaxis]
lambda0, lambda1, lambda2 = 2, 2, 0

# Opstiller dekomponering
X = np.column_stack((v0/np.linalg.norm(v0,2), v1/np.linalg.norm(v1,2), v2/np.linalg.norm(v2,2), v3/np.linalg.norm(v3,2)))
D = np.diag((lambda0, lambda1, lambda2, lambda3))

# Tjekker om det passer
machine_epsilon = np.finfo(float).eps
np.allclose(A, X @ D @ X.T, rtol=machine_epsilon, atol=machine_epsilon)
````
