# 6.1



## (a)



$\kappa(x) = \frac{|x||-\frac{2x}{(1+x^2)^2}|}{|\frac{1}{1+x^2}|} = \frac{|-\frac{2x^2}{(1+x^2)^2}|}{|\frac{1}{1+x^2}|} = |-\frac{2x^2}{1+x^2}| = |\frac{2x^2}{1+x^2}| = \frac{2x^2}{1+x^2} \in [0,2)$

```python
x = np.linspace(-100, 100, 1000)
k = np.abs(x)*np.abs(-(2*x)/((1+x**2)**2))/(np.abs(1/(1+x**2)))

fig, ax = plt.subplots()

ax.plot(x, k)
```

## (b)



$\kappa(x) = \frac{|x||cos(x)|}{|sin(x)|}$

Denne eksploderer ved $x = n\pi, n \in \mathbb{Z}$,

```python
x = np.linspace(-15, 15, 1000)
k = np.abs(x)*np.abs(np.cos(x))/np.abs(np.sin(x))

fig, ax = plt.subplots()

ax.plot(x, k)
```

## (c)



$$\kappa(x) = \frac{||\begin{bmatrix} x \\ y \end{bmatrix}||_2 ||\begin{bmatrix} y \\ x \end{bmatrix}||_2}{|f(x)|}$$

$$= \frac{\sqrt{x^2 + y^2} * \sqrt{y^2 + x^2}}{|xy|}$$

$$= \frac{x^2 + y^2}{|xy|}$$

$$= |\frac{x}{y}| + |\frac{y}{x}|$$

Når $x$ eller $y$ er tæt på nul, så får vi stor relativ fejl. Det samme sker hvis enten $x$ er meget større end $y$, eller omvendt, da vi så vil få at en af brøkerne bliver større.

## (c)



```python
n = 361
t = np.linspace(0, 2*np.pi, n)
vinkel = np.empty(n)
for i in range(n):
    w0 = np.array([np.cos(t[i]), np.sin(t[i])])[:,np.newaxis]
    w1 = np.array([-np.sin(t[i]), np.cos(t[i])])[:,np.newaxis]
    Aw0 = A @ w0
    Aw1 = A @ w1
    vinkel[i] = np.degrees(np.arccos(np.vdot(Aw0, Aw1)
                        / np.linalg.norm(Aw0)
                        / np.linalg.norm(Aw1)))

fig, ax = plt.subplots()
ax.set_xticks([0, 90, 180, 270, 360])
ax.plot(np.degrees(t), vinkel)
ax.grid()
```

## (a)



For en diagonal matrix vil vi have at $U$'s søjler er en samling af $e_i$, og $V^T$'s rækker er de tilsvarende $e_i$. Singulærværdierne vil så være de sorterede værdier på diagonalen af $A$.

```python
u = np.array([[0, 1, 0],
               [1, 0, 0],
               [0, 0, 1]])

sigma =np.array([[3, 0, 0],
               [0, 2, 0],
               [0, 0, 1]])

u @ sigma @ u
```

## (b)



For en diagonal matrix vil vi have at $U$'s søjler er en samling af $e_i$, og $V^T$'s rækker er de tilsvarende $e_i$. Singulærværdierne vil så være de sorterede absolutte værdier på diagonalen af $A$.

En lille note er at de negative indgange på diagonalen vil have lavet den tilhørende søjle i $U$ eller tilhørende søjle i $V^T$ negativ, for at holde styr på fortegnet.

## (c)



$A^T = (U \Sigma V^T)^T = V \Sigma^T U^T = V \Sigma U^T$

## (d)



$U = A$, $\Sigma = I_n$, $V^T = I_n$

## (e)



Alle singulærværdier er $> 0$

Fordi $\det(A) = \prod_\Sigma \sigma$, og hvis blot en af singulærværdierne er $0$, så bliver determinanten $0$, og så er matricen $A$ ikke invertibel.

## (f)



Når de singulærværdier er forskellige er singulærvektorerne entydigt
bestemt op til multiplikation med ±1, sammenlign med beviset for eksistens
af SVD. Når der er singulærværdier, der gentages, er der større frihed, da der
kan roteres rundt i rummet udspændt af de singulærvektorer med samme
singulærværdi.

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]], dtype=float)

u, s, vt = np.linalg.svd(A, full_matrices=False)

u, s, vt
```
