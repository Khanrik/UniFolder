```ocaml
let y = "(fun g. (fun y. g @ (y @ y)) @ (fun z. g @ (z @ z)))";;
let fac = "(fun f x. ? @ (= @ x @ 0) @ 1 @ (* @ x @ (f @ (- @ x @ 1))))";;
let yfac = y ^ " @ " ^ fac;;
dbparser yfac
```

```ocaml
let yfac3 = (yfac ^ " @ " ^ "3");;
dbparser yfac3
```

```ocaml
db_eval_int_lazy yfac3;;
```

```ocaml
let num = "(fun f x. ? @ (= @ x @ 0) @ Z @ (S @ (f @ (- @ x @ 1))))";;
let ynum3 = y ^ " @ " ^ num ^ " @ 3"
```

```ocaml
db_eval_int_lazy ynum3
```
