### Problem 1
$\frac{dx}{dt} = x - x^{2}, x(0) = 2$

Find general solution
\begin{align*}
\frac{dx}{dt} &= x - x^{2}\\
\frac{1}{x-x^{2}} dx &= 1 dt\\
\int \frac{1}{x(1-x)}  \, dx  &= \int 1 \, dt\\
\int \frac{1}{x} + \frac{1}{1-x} \, dx &= t + C_{1}\\
\ln(x) - \ln(1-x) &= t + C_{1}\\
e^{\ln(x) - \ln(1-x)} &= t + C_{1}\\
\frac{x}{1-x} &= e^{t + C_{1}}\\
x &= (C_{2}e^{t}) \cdot (1-x) \qquad C_{2} = e^{C_{1}}\\
x &= C_{2}e^{t} - C_{2}e^{t}x\\
x + C_{2}e^{t}x &= C_{2}e^{t}\\
x (1 + C_{2}e^{t}) &= C_{2}e^{t}\\
x &= \frac{C_{2}e^{t}}{1 + C_{2}e^{t}}
\end{align*}

Insert initial condition
\begin{align*}
2 &= \frac{C_{2}e^{0}}{1 + C_{2}e^{0}}\\
2 &= \frac{C_{2}}{1 + C_{2}}\\
2 (1 + C_{2}) &= C_{2}\\
2 + 2C_{2} &= C_{2}\\
2 &= -C_{2}\\
C_{2} &= -2
\end{align*}

Insert integration constant
\begin{align*}
x &= -\frac{2e^{t}}{1 - 2e^{t}}\\
x &= \frac{2e^{t}}{2e^{t}-1}
\end{align*}

```python
#Define ODE
def df(x,y):
    return y - y**2

#Show scipy numerical solution True/False
show_solution = True

#Define boundaries of x and y in the slope field plot
minx = -5.
maxx = 15.
miny = -10.
maxy = 10.

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Define the initial value(s) - Even if only one: Make a 1D array
x0 = 0
y0 = np.array([1.1, 0.1, 5])

#Create and normalize direction vectors
u = np.ones(x.shape)
v = df(x,y)
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
ax.set_title("Slope Field")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()
# plt.savefig("slope_field")

# solution
t = np.linspace(minx, maxx, 100)
x = (2 * np.exp(t))/(2 * np.exp(t) - 1)
x1 = (4 * np.exp(t))/(4 * np.exp(t) - 1)
x2 = (np.exp(t))/(np.exp(t) - 1)

plt.plot(t, x, color="red")
plt.plot(t, x1, color="gray")
plt.plot(t, x2, color="gray")
plt.show()
```

### Problem 2
$\frac{dx}{dt} = 10x - x^{2}, x(0) = 1$

Find general solution
\begin{align*}
\frac{dx}{dt} &= 10x - x^{2}\\
\int \frac{1}{x(10-x)} \, dx &= \int 1 \, dt\\
\int \frac{1}{x} + \frac{1}{10-x} \, dx &= t + C_{1}\\
\ln(x) - \ln(10-x) &= t + C_{1}\\
e^{\ln(x) - \ln(10-x)} &= t + C_{1}\\
\frac{x}{10-x} &= e^{t + C_{1}}\\
x &= (C_{2}e^{t}) \cdot (10-x) \qquad C_{2} = e^{C_{1}}\\
x &= 10C_{2}e^{t} - C_{2}e^{t}x\\
x + C_{2}e^{t}x &= 10C_{2}e^{t}\\
x (1 + C_{2}e^{t}) &= 10C_{2}e^{t}\\
x &= \frac{10C_{2}e^{t}}{1 + C_{2}e^{t}}
\end{align*}

Insert initial value
\begin{align*}
1 &= \frac{10C_{2}e^{0}}{1 + C_{2}e^{0}}\\
1 &= \frac{10C_{2}}{1 + C_{2}}\\
1 + C_{2} &= 10C_{2}\\
1 &= 9C_{2}\\
C_{2} &= \frac{1}{9}
\end{align*}

Insert integration constant
$$
x = \frac{\frac{10}{9} e^{t}}{1 + \frac{1}{9}e^{t}}
$$

```python
#Define ODE
def df(x,y):
    return 10 * y - y**2

#Show scipy numerical solution True/False
show_solution = True

#Define boundaries of x and y in the slope field plot
minx = -10.
maxx = 15.
miny = -5.
maxy = 15.

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Define the initial value(s) - Even if only one: Make a 1D array
x0 = 0
y0 = np.array([1.1, 0.1, 5])

#Create and normalize direction vectors
u = np.ones(x.shape)
v = df(x,y)
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
ax.set_title("Slope Field")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()
# plt.savefig("slope_field")

# solution
t = np.linspace(minx, maxx, 100)
x = (10.0/9 * np.exp(t))/(1 + 1.0/9 * np.exp(t))
x1 = (10.0/6 * np.exp(t))/(1 + 1.0/6 * np.exp(t))
x2 = (10.0/12 * np.exp(t))/(1 + 1.0/12 * np.exp(t))

plt.plot(t, x, color="red")
plt.plot(t, x1, color="gray")
plt.plot(t, x2, color="gray")
plt.show()
```
