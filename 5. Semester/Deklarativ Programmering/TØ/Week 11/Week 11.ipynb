### a. $(x = S(y) \Rightarrow y = z \land \text{in}(x, A(z)))[S(y)/x, z + x/y, y/z]$

\begin{align*}
&(x = S(y) \Rightarrow y = z \land \text{in}(x, A(z)))[S(y)/x, z + x/y, y/z]\\
&= (x = S(y))[S(y)/x, z + x/y, y/z] \Rightarrow (y = z)[S(y)/x, z + x/y, y/z] \land (\text{in}(x, A(z)))[S(y)/x, z + x/y, y/z]\\
&= x[S(y)/x, z + x/y, y/z] = S(y)[S(y)/x, z + x/y, y/z] \Rightarrow y[S(y)/x, z + x/y, y/z] = z[S(y)/x, z + x/y, y/z] \land \text{in}(x[S(y)/x, z + x/y, y/z], A(z)[S(y)/x, z + x/y, y/z])\\

&= S(y) = S(z + x) \Rightarrow z + x = y \land \text{in}(S(y), A(y))
\end{align*}

## (4) Calculate:

### a. $\Delta \{ (f(x,x), f(x, f(x))) \} \emptyset$

\begin{align*}
&\Delta \{ (f(x,x), f(x, f(x))) \} \emptyset\\
&= \Delta \{ ((x,x), (x, f(x))) \} \emptyset\\
&= \Delta \{ (x, f(x)) \} \emptyset\\
&= \textbf{fail}
\end{align*}

We have: $f(x,x)$ does not match $f(x, f(x))$

### b. $\Delta \{ ((k+p) * A(u), (1 + G(2,i)) * A(7 * i)) \} \emptyset$

\begin{align*}
&\Delta \{ ((k+p) * A(u), (1 + G(2,i)) * A(7 * i)) \} \emptyset\\
&= \Delta \{ (k+p, 1 + G(2,i)), (A(u), A(7 * i)) \} \emptyset\\
&= \Delta \{ (k, 1), (p, G(2,i)), (A(u), A(7 * i)) \}\emptyset\\
&= \Delta \{ (p, G(2, i)), (A(u), A(7 * i)) \} \{ (k, 1) \}\\
&= \Delta \{ (A(u), A(7 * i)) \} \{ (k, 1), (p, G(2, i)) \}\\
&= \Delta \{ (u, 7 * i) \} \{ (k, 1), (p, G(2, i)) \}\\
&= \Delta \emptyset \{ (k, 1), (p, G(2, i)), (u, 7 * i) \}\\
&= \{ (k, 1), (p, G(2, i)), (u, 7 * i) \}
\end{align*}

We have: $(k+p) * A(u)[k/1, p/G(2,i), u/7 * i] = (1 + G(2,i)) * A(7 * i)$

### c. $\Delta \{ (h(x,y) + x, h(S(0), x) + S(0)) \} \emptyset$

\begin{align*}
&\Delta \{ (h(x,y) + x, h(S(0), x) + S(0)) \} \emptyset\\
&= \Delta \{ (h(x,y), h(S(0), x)), (x, S(0)) \} \emptyset\\
&= \Delta \{ ((x,y), (S(0), x)), (x, S(0)) \} \emptyset\\
&= \Delta \{ (x, S(0)), (y, x), (x, S(0)) \} \emptyset\\
&= \Delta \{ (y, x), (x, S(0)) \} \{ (x, S(0)) \}\\
&= \Delta \{ (y, x) \} \{ (x, S(0)) \}\\
&= \Delta \emptyset \{ (x, S(0)), (y, x) \}
\end{align*}

We have: $h(x,y) + x [S(0)/x, x/y] = h(S(0), x) + S(0)$
