### (a) $x[n] = \left( \frac{1}{2} \right)^{n} (u[n] - u[n - 10])$

\begin{align*}
X[z] &= \sum_{n = -\infty}^{\infty} \left( \frac{1}{2}  \right)^{n} (u[n] - u[n-10]) z^{-n}\\
&= \sum_{n = 0}^{9} \left( \frac{1}{2}  \right)^{n} z^{-n}\\
&= \frac{1 - \left( \frac{1}{2}z^{-1} \right)^{10}}{1 - \frac{1}{2}z^{-1}}\\
&= \frac{1 - \frac{1}{2^{10}} z^{-10}}{1 - \frac{1}{2}z^{-1}}\\
&= \frac{z^{10} - \frac{1}{2^{10}} z^{0}}{z^{10} - \frac{1}{2}z^{9}}\\
&= \frac{z^{10} - \frac{1}{2^{10}}}{z^{9} (z - \frac{1}{2})}
\end{align*}

**Poles**

The roots of the denominator $\frac{z^{10} - \frac{1}{2^{10}}}{z^{9} (z - \frac{1}{2})}$ are $z=0$ (multiplicity 9) and $z=\frac{1}{2}$.

**Zeros**

The roots of the numerator $z^{10} - \frac{1}{2^{10}}$ is $z = \frac{1}{2}$ with a multiplicity of 10. The denominator however also had this exact root, meaning that the cancel out leaving the zeros to be $z=\frac{1}{2}$ and poles to be $z=0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the zeros (10 zeros evenly spaced on a circle of radius 1/2)
N = 10  # Number of zeros
zero_radius = 1/2
zeros = zero_radius * np.exp(1j * (2 * np.pi * np.arange(N) / N))

# Define the poles (one at z = 1/2 and nine at z = 0)
poles = np.array([1/2] + [0]*9)  

# Plot setup
fig, ax = plt.subplots(figsize=(6,6))
ax.axhline(0, color='black', linewidth=1)  # X-axis
ax.axvline(0, color='black', linewidth=1)  # Y-axis
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_xlabel("Re{z}")
ax.set_ylabel("Im{z}")
ax.set_title("Pole-Zero Plot")

# Plot unit circle
theta = np.linspace(0, 2*np.pi, 300)
ax.plot(np.cos(theta), np.sin(theta), 'k--', linewidth=1, label="Unit Circle")

# Plot zeros (blue circles)
ax.scatter(zeros.real, zeros.imag, marker='o', color='b', s=100, label="Zeros")

# Plot poles (red crosses)
ax.scatter(poles.real, poles.imag, marker='x', color='r', s=100, label="Poles")

# Legend and grid
ax.legend()
ax.grid(True, linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```
