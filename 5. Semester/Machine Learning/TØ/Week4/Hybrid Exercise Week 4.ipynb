## 3) A set of 2-dimensional data points x ∈ R2 follow the distribution in Eq. (1):
$$
p(x) = N (x|μ, Σ) = 1 (2π)D/2|Σ|1/2 exp \{ − 1 2 (x − μ)T Σ−1(x − μ) \}
$$
The parameters μ ∈ R2 and Σ =∈ R2×2 of the distribution are equal to:
$$
μ = 
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$
and
$$
\Sigma = 
\begin{bmatrix}
1 & 0 \\
0 & 2
\end{bmatrix}
$$
Answer the following questions:

Given three data points $x_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ , $x_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$ and $x_3 = \begin{bmatrix} 0.5 \\ 0.5 \end{bmatrix}$ , add the (in)equality signs in the following, and describe why these signs are used (provide a qualitative explanation and do not calculate the values of $p(x)$ for the three points):
\begin{align*}
p(x_1) &\qquad p(x_2) \\
p(x_1) &\qquad p(x_3) \\
p(x_3) &\qquad p(x_2)
\end{align*}



The fraction on the left side of the expression is simply a factor that is multiplied on all three points, so it doesn't affect the inequalities. The exponent part is what determines the differences. Furthermore, $\mu$ is the 0 vector and $\frac{1}{2}$ is a constant factor, meaning that they can be ignored in finding the inequality. Since $\exp(-x)$ is a decreasing function, the larger the value of $x^T \Sigma^{-1} x$, the smaller the probability $p(x)$. Inserting the values of of $\Sigma^{-1}$ the expressions becomes:
$$
x_1 \cdot 1 \cdot x_1 + x_2 \cdot \frac{1}{2} \cdot x_2 = x_1^2 + \frac{1}{2} x_2^2
$$

Given this relation it is obvious that $p(x_1)$ has to be smallest since the value of the above expression would be 1. For $x_2$ the expression would be $\frac{1}{2}$, and for $x_3$ it would be $0.5^2 + \frac{1}{2} \cdot 0.5^2 = \frac{3}{8}$. Therefore, the inequalities are:
\begin{align*}
p(x_1) &< p(x_2) \\
p(x_1) &< p(x_3) \\
p(x_3) &> p(x_2)
\end{align*}
