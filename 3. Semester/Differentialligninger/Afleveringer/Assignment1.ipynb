## Problem 1
Consider the differential equation $dy/dx = (y-3)^{2}$.

### 1.
Find the general solution of the differential equation and the singular solution that is not included in the general solution.

First the general solution is found by solving it as a seperable differential equation:

\begin{align*}
\frac{dy}{dx} &= (y-3)^{2}\\
\frac{1}{(y-3)^{2}}dy &= 1 dx\\
\int \frac{1}{(y-3)^{2}} \, dy &= \int 1 \, dx\\
- \frac{1}{y-3} &= x + C\\
y - 3 &= - \frac{1}{x+C}\\
y &= 3 - \frac{1}{x + C} &\text{(the general solution)}
\end{align*}

It can now be seen that since the right hand side of the general solution features 3 subtracted by a fraction. Since only denominator of the fraction contains variables and the numerator is not 0, the fraction will never be 0 and the solution will never reach

$$
y = 3 - 0
$$

This means that the singular solution that is not included in the general solution, which was found, is $y = 3$.

### 2.

Since $y(0) \neq 3$, the inital value problem can be solved using the general solution from before. This is done by substituting $y$ for 4 and $x$ for 0.

\begin{align*}
4 &= 3 - \frac{1}{0 + C}\\
1 &= - \frac{1}{C}\\
C &= -1
\end{align*}

The solution for the intial value is then
$$
y = 3 - \frac{1}{x - 1}
$$

```python
#Define ODE
def df(x,y):
    return (y-3)**2

def sol(x,c):
    return 3 - 1/(x + c)

#Define boundaries of x and y in the slope field plot
minx = -4.
maxx = 6.
miny = -4.
maxy = 10.

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Create and normalize direction vectors
u = np.ones(x.shape)
v = df(x, y)
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()

x_sol = np.linspace(minx, maxx, 1000)
c = -1
y_sol = sol(x_sol, c)

ax.plot(x_sol, y_sol, color='red')
```

The solution consists of two parts, which are disconnected by an asymptote at $x=1$. This disconnection is due to the fact that the denominator in fraction of the solution would become 0 at $x=1$. The solution function is therefore not defined there. When $x$ approaches $-\infty$ the solution approaches $y = 3$, but it will never be reached since it is not defined for the general solution that was used. Toward $+\infty$ the function of the solution also approaches $y=3$, but since the inital value problem was for $y(0)$, the solution only refers to the left side of asymptote and not the right side, which is a solution for a different initial value problem.

### 3.
What does Theorem 1, Existence and Uniqueness of Solutions say about the solution(s) of the initial value problem $y' = (y − 3)^{2}, y(a) = b$ for arbitrary choices of $a$ and $b$?

The theorem states that for some open interval $I$ there is one and only one solution for an initial value problem. Since we have found a solution that is not defined for x=1, the solution is defined only on the open interval $I = (-\infty, 1)$. This means that the right side of the function is not part of the solution, but that it is a solution to a different initial value problem, which is defined on the open interval $I = (1, \infty)$. This can be generalized by seeing that for any choice of $a$ and $b$, the constant $C$ can be described as

\begin{align*}
b &= 3 - \frac{1}{a + C}\\
b-3 &= - \frac{1}{a + C}\\
(a + C)(b - 3) &= -1\\
a + C &= - \frac{1}{b - 3}\\
C &= \frac{1}{3 - b} - a
\end{align*}

The denominator of the solution would then approach 0 as $x$ approaches $\frac{1}{b - 3} + a$, which means that there will be an asymptote there. Like before the same function would then be two different solution on the open intervals $I = (-\infty, \frac{1}{b - 3} + a)$ and $I = (\frac{1}{b - 3} + a, \infty)$ respectively.

There is still the special case for $y(a) = 3$, which is not defined for the general solution. Here there is only a single function, $y(x) = 3$, counting as a single solution to the problem defined for all values of $x$ in the open interval $I = (-\infty, \infty)$.

## Problem 2
Consider the differential equation $y' = (y − 2) \cos(x)$.

### 1. 
Solve the differential equation by hand for the initial value $y(0) = 0$. Account for all intermediate steps in the solution.

First the general solution is found by solving it as a seperable differential equation:

\begin{align*}
\frac{dy}{dx} &= (y − 2) \cos(x)\\
\frac{1}{y-2} dy &= \cos(x) dx\\
\int \frac{1}{y-2} \, dy &= \int \cos(x) \, dx
\end{align*}

Here substitution can be used to solve the integral on the left side, but it can also simply be solved by knowing the derivative $\frac{d}{dx} \ln(f(x)) = \frac{1}{f(x)} \cdot f'(x)$. Since the $\frac{d}{dy} y-2 = 1$, the integral of $y-2$ simply becomes $\ln(y-2)$. 

\begin{align*}
\ln(y-2) + C_{y} &= \sin(x) + C_{x}\\
\ln(y-2) &= \sin(x) + C_{1} &C_{1} = C_{x} - C_{y}\\
y-2 &= e^{\sin(x) + C_{1}}\\
y-2 &= e^{\sin(x)} \cdot e^{C_{1}}\\
y-2 &= C_{2} e^{\sin(x)} &C_{2} = e^{C_{1}}\\
y &= C_{2} e^{\sin(x)} + 2 &\text{(the general solution)}
\end{align*}

The initial value can then be inserted into the general solution to find $C_{2}$

\begin{align*}
0 &= C_{2} e^{\sin(0)} + 2\\
-2 &= C_{2} e^{0}\\
C_{2} &= -2
\end{align*}

Inserting the value found for $C_{2}$ back into the general solution gives the particular solution for the initial value problem

$$
y = -2 e^{\sin(x)} + 2 \qquad \text{(the particular solution)}
$$



### 2. 
Discuss the solutions to the differential equation if other initial values than 0 are used, i.e., the $y(0) = b, b \neq 0$ case. Include plots of the slope field and representative solutions.

As can be seen when finding $C_{2}$ in the previous part, it only depends on the value of $b$ since $e^{\sin(x)} = 1$. For any value $b$ the constant $C_{2}$ can be found as
$$
C_{2} = b - 2
$$
making it very easy to find solutions using for loops in python.

This also means that the solution for the initial value $y(0) = 2$, would become the constant function $y = (2-2)e^{\sin(x)} + 2 = 2$. Every other function however will grow and shrink exponentially with a periodicity defined by the sinus function in the exponent. Various solutions (excluding $y(0)=0$) can be seen below.

```python
#Define ODE
def df(x,y):
    return (y-2) * np.cos(x)

def sol(x,c):
    return c * np.exp(np.sin(x)) + 2

#Show scipy numerical solution True/False
show_solution = True

#Define boundaries of x and y in the slope field plot
minx = -10.
maxx = 10.
miny = -12.5
maxy = 12.5

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Define the initial value(s) - Even if only one: Make a 1D array
x0 = minx
y0_values = np.array([-2, -1, 1, 2, 3, 4, 5])

#Create and normalize direction vectors
u = np.ones(x.shape)
v = df(x,y)
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
if show_solution:
    for y0 in y0_values:
        #Solution
        xsol = np.linspace(x0, maxx, xsteps*100)
        c = y0-2
        ysol = sol(xsol, c)
        ax.plot(xsol, ysol, label=f'$y(0)$ = {y0}')
        ax.legend(loc='upper left', fontsize=12)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()
# plt.savefig("slope_field")
plt.show()
```

## Problem 3
Consider the differential equation $y' = 2x − y$.

### 1.
Solve the differential equation with the initial value $y(0) = 3$.

First the general solution is found. This is done by getting the equation on the form $dy/dx + P(x)y = Q(x)$:

\begin{align*}
\frac{dy}{dx} &= 2x - y\\
\frac{dy}{dx} + y &= 2x\\
\end{align*}

The general solution can now be found using Theorem 1.5.1:

\begin{align*}
y(x) &= e^{-\int 1 \, dx } \left[ \int \left( 2x e^{\int 1 \, dx } \right) \, dx + C \right]\\
&= e^{-x} \left[ \int 2xe^{x} \, dx + C \right]\\
&= e^{-x} \left[ 2x \cdot e^{x} - \int 2e^{x} \, dx + C \right] &\text{(integration by parts)}\\
&= e^{-x} [2x \cdot e^{x} - 2e^{x} + C]\\
&= 2x - 2 + Ce^{-x}\\
&= Ce^{-x} + 2x - 2 &\text{(the general solution)}
\end{align*}

The constant $C$ corresponding to the problem can be found by inserting the inital value:

\begin{align*}
3 &= C e^{-0} + 2 \cdot 0 - 2\\
3 &= C - 2\\
C &= 5
\end{align*}

Inserting the value found for $C$ back into the general solution gives the particular solution for the initial value problem

$$
y = 5e^{-x} + 2x - 2 \qquad \text{(the particular solution)}
$$

### 2.
Account for the asymptotic behaviour of the solution for this initial value as well as other initial values. Include any necessary plots.

The solution will be plotted below as well as the solutions with odd values of $C$ from -5 to 3.

```python
#Define ODE
def df(x,y):
    return 2*x - y

def sol(x,c):
    return c * np.exp(-x) + 2*x - 2

#Show scipy numerical solution True/False
show_solution = True

#Define boundaries of x and y in the slope field plot
minx = -10.
maxx = 20.
miny = -30
maxy = 30

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Define the initial value(s) - Even if only one: Make a 1D array
x0 = minx
c_values = np.array([-5, -3, -1, 1, 3, 5])

#Create and normalize direction vectors
u = np.ones(x.shape)
v = df(x,y)
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
if show_solution:
    for c in c_values:
        #Solution
        xsol = np.linspace(x0, maxx, xsteps*100)
        ysol = sol(xsol, c)
        ax.plot(xsol, ysol, label=f'$y = {c} e^{{-x}} + 2x - 2$')
        ax.legend(loc='lower right', fontsize=12)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()
plt.show()
```

As can be seen in the graph above all the solutions have a vertical asymptote, which is shifted further to the right along the x-axis the larger the value of $C$ is numerically. This makes sense, since $Ce^{-x}$ will approach $\infty$ as $x$ approaches $-\infty$. Something all solutions have in common is a asymptote defined by the function $y(x) = 2x - 2$. They all approach this function since the $Ce^{-x}$ term approaches $0$ as $x$ approaches $\infty$, making the entire expression of any solution approach $y = 2x - 2$.

### 3.
Assume that the differential equation has been solved for two different initial values, $y_{1}(a) = b_{1}$ and $y_{2}(a) = b_{2}$ with $b_{1} \neq b_{2}$.

Can the two solutions, $y_{1}(x)$ and $y_{2}(x)$, cross each other?

Based on the general solution we can let $y_{1}(x) = C_{1}e^{-x} + 2x - 2$ and $y_{2}(x) = C_{2}e^{-x} + 2x - 2$, where $C_{1} \neq C_{2}$ since they are solutions to different initial values. We can set them to be equal to each other to see if they cross and if so for which value of $x$.

\begin{align*}
C_{1}e^{-x} + 2x - 2 &= C_{2}e^{-x} + 2x - 2\\
C_{1} &= C_{2}
\end{align*}

Every term other than the constants cancel out leaving the expression as a contradiction, since they cannot be equal if they are solutions to different initial values. Thus, the two solutions will never cross.

## Problem 4
Consider the logistic differential equation $y' = 2y(5 − y)$.

### 1.
Compute and plot the solutions for $y(0) = 1$ and $y(1) = 1$. What features do the two solutions have in common?

The general solution is first calculated as a seperable differential equation:

\begin{align*}
\frac{dy}{dx} &= 2y(5 - y)\\
\int \frac{1}{2y(5 - y)} \, dy &= \int 1 \, dx\\
\frac{1}{2} \int \frac{1}{y(5 - y)} \, dy &= x + C_{1}
\end{align*}

The left side can be rewritten using partial fractions as follows

\begin{align*}
\frac{A}{y} + \frac{B}{5-y} &= \frac{1}{y(5-y)}\\
\frac{A \cdot y(5 - y)}{y} + \frac{B \cdot y(5 - y)}{5 - y} &= 1\\
A \cdot (5 - y) + B \cdot y &= 1\\
A \cdot 5 + B \cdot 0 &= 1 \qquad y=0 \text{ is chosen to find A since any y value is valid}\\
A &= \frac{1}{5}\\
A \cdot 0 + B \cdot 5 &= 1 \qquad y=5 \text{ is chosen to find B}\\
B &= \frac{1}{5}
\end{align*}

Inserting the partial fractions back in the caluculation yields

\begin{align*}
\frac{1}{2} \int \frac{1}{5y} + \frac{1}{5 (5 - y)} \, dy &= x + C_{1}\\
\frac{1}{10} \int \frac{1}{y} + \frac{1}{5 - y} \, dy &= x + C_{1}\\
\frac{1}{10} \cdot (\ln(y) - \ln(5 - y)) &= x + C_{1}\\
\ln(y) - \ln(5 - y) &= 10x + 10C_{1}\\
e^{\ln(y) - \ln(5 - y)} &= e^{10x + 10C_{1}}\\
\frac{y}{5-y} &= e^{10x} \cdot e^{10C_{1}}\\
y &= e^{10x} \cdot e^{10C_{1}} \cdot (5 - y) &C_{2} = e^{10C_{1}}\\
y &= e^{10x} \cdot C_{2} \cdot 5 - e^{10x} \cdot C_{2} \cdot y\\
y + e^{10x} \cdot C_{2} \cdot y &= 5 \cdot e^{10x} \cdot C_{2}\\
y \cdot (1 + e^{10x} \cdot C_{2}) &= 5 \cdot e^{10x} \cdot C_{2}\\
y &= \frac{5 \cdot e^{10x} \cdot C_{2}}{1 + e^{10x} \cdot C_{2}}\\
y &= \frac{5 \cdot e^{10x} \cdot \frac{1}{C_{3}}}{1 + e^{10x} \cdot \frac{1}{C_{3}}} &C_{3} = \frac{1}{C_{2}}\\
y &= \frac{5 \cdot e^{10x}}{C_{3} + e^{10x}} &\text{(the general solution)}
\end{align*}

The initial value $y(0) = 1$ can now be inserted into the general solution:

\begin{align*}
1 &= \frac{5 \cdot e^{10 \cdot 0}}{C_{3} + e^{10 \cdot 0}}\\
C_{3} + 1 &= 5\\
C_{3} &= 4
\end{align*}

This means that the solution for the initial value $y(0) = 1$ is $y_{0} = \frac{5 \cdot e^{10x}}{4 + e^{10x}}$.

Now inserting the initial value $y(1) = 1$ yields:

\begin{align*}
1 &= \frac{5 \cdot e^{10 \cdot 1}}{C_{3} + e^{10 \cdot 1}}\\
C_{3} + e^{10} &= 5 \cdot e^{10}\\
C_{3} &= 5 \cdot e^{10} - e^{10}\\
C_{3} &= 4 \cdot e^{10}
\end{align*}

Giving the solution $y_{1} = \frac{5 \cdot e^{10x}}{4 \cdot e^{10} + e^{10x}}$

These two solutions can now be plotted in python


```python
#Define ODE
def df(x,y):
    return 2 * y * (5 - y)

def sol(x,c):
    return (5 * np.exp(10 * x)) / (c + np.exp(10 * x))

#Define boundaries of x and y in the slope field plot
minx = -10.
maxx = 10.
miny = -4.
maxy = 7.

#Define step n along each axis (resolution of slope field)
xsteps = 30
ysteps = 30

#Create linspace from parameters 
xs = np.linspace(minx, maxx, xsteps)
ys = np.linspace(miny, maxy, ysteps)

#Create a meshgrid from the linspaces
x, y = np.meshgrid(xs, ys)

#Create and normalize direction vectors
u = np.ones(x.shape)
v = sol(x,y)
norm = np.sqrt(u**2 + v**2)
u /= norm
v /= norm

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.quiver(x, y, u, v, angles='xy', headwidth=2)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()

x_sol = np.linspace(minx, maxx, 1000)
c_0 = 4
y_0 = sol(x_sol, c_0)
ax.plot(x_sol, y_0, color='red', label="y(0) = 1")

c_1 = 4 * np.exp(10)
y_1 = sol(x_sol, c_1)
ax.plot(x_sol, y_1, color='blue', label="y(1) = 1")
ax.legend(loc='upper left', fontsize=12)
```

Both functions have the same sigmoid shape, but that the solution for $y(1) = 1$ is shifted to the right. It makes sense that the functions approach 0 toward $-\infty$ since $e^{10x}$ approaches 0 and the entire expression would approach $\frac{0}{C}$. When approaching $+\infty$, the constant in the denominator becomes obsolete, meaning the expression approaches $\frac{5 e^{10x}}{e^{10x}}$. The two $e^{10x}$ cancels out leaving 5, which is also seen in the graph. The reason for the shift is that the constant in the denominator does make a difference at small x-values, meaning that it takes a longer time before the $e^{10x}$ terms explode the bigger the constant is.

### 3.
Assume that we use the logistic differential equation to model a population and that an additional term is added to the equation to model harvesting, i.e, $y' = 2y(5 − y) − h, h > 0$.

What is the maximum value of $h$ for which the population still exists?

For the population to still exist, the harvest must either make the population grow (positive slope) or stay at a constant (slope of 0). It can be seen that the growth of the population is modelled after a quadratic equation depending on the current population ($y$) and the harvesting ($h$)
$$
\frac{dy}{dx} = 2y(5-y) = -2y^{2} + 10y - h
$$

Since the first term ($-2y^{2}$) is negative, it must mean that the vertex of the model is also the maximum point. The constant $h$ will move the vertex toward $-\infty$ as $h$ approaches $\infty$. When the vertex gets below $\frac{dy}{dx} = 0$, it means that the population can no longer grow nor stay at a constant, meaning that it will cease to exist. The maximal value of $h$ is then the value that places the vertex at exactly $\frac{dy}{dx} = 0$. In other words the graph of the equation must only have a single root, which is true if the discriminant of the equation is 0. The discriminant can then be set to 0 and the resulting equation can be solved for $h$.

\begin{align*}
0 &= b^{2} - 4ac\\
0 &= 10^{2} - 4 \cdot (-2) \cdot (-h)\\
0 &= 100 - 8h\\
h &= \frac{100}{8} = 12.5
\end{align*}

For the population to still exist, the harvesting constant can have a maximal value of $h = 12.5$.

## Problem 5
Consider a system modelled by the differential equation
$$
\frac{dy}{dx} = M(x) − 2y, \; M(x) =
\begin{cases}
5, & 0 \leq x \leq 5 \\
0, & x > 5
\end{cases}
$$

### 1. 
Solve the differential equation for the initial value $y(0) = 1$ and plot the solution for $0 < x < 10$.

First the general solution is found. This is done by getting the equation on the form $dy/dx + P(x)y = Q(x)$:

\begin{align*}
\frac{dy}{dx} &= M(x) - 2y\\
\frac{dy}{dx} + 2y &= M(x)\\
\end{align*}

The general solution can now be found using Theorem 1.5.1. This is first done for the interval $0 \leq x \leq 5$:

\begin{align*}
y_{1}(x) &= e^{-\int 2 \, dx } \left[ \int \left( 5 e^{\int 2 \, dx } \right) \, dx + C_{1} \right]\\
&= e^{-2x} \left[ 5 \int e^{2x} \, dx + C_{1} \right]\\
&= e^{-2x} \left[  5 \cdot \frac{1}{2} \cdot e^{2x} + C_{1} \right]\\
&= \frac{5}{2} + C_{1} e^{-2x} &\text{(general solution for 0 ≤ x ≤ 5)}
\end{align*}

It is then found for for $5 < x < 10$

\begin{align*}
y_{2}(x) &= e^{-\int 2 \, dx } \left[ \int \left( 0 e^{\int 2 \, dx } \right) \, dx + C_{2} \right]\\
&= e^{-2x} \left[ C_{2} \right]\\
&= C_{2} e^{-2x} &\text{(general solution for 5 < x < 10)}
\end{align*}

The initial value is then inserted into the first general solution:

\begin{align*}
1 &= \frac{5}{2} + C_{1} e^{-2 \cdot 0}\\
C_{1} &= -\frac{3}{2}
\end{align*}

The particular solution for the interval $0 \leq x \leq 5$ is therefore:
$$
y(x) = \frac{5}{2} - \frac{3}{2} e^{-2 \cdot x}
$$

To ensure the continuity between the part before and after $x=5$, the particular solutioon $y_{1}(x)$ can be set equal to $y_{2}(x)$ at $x=5$. The particular solution for the interval $5 \leq x \leq 10$ is then found by isolating $C_{2}$:

\begin{align*}
C_{2}e^{-2 \cdot 5} &= \frac{5}{2} - \frac{3}{2} e^{-2 \cdot 5}\\
C_{2} &= \frac{5}{2}e^{10} - \frac{3}{2}
\end{align*}

The new constant is then inserted into $y_{2}(x)$:

\begin{align*}
y_{2}(x) &= (\frac{5}{2}e^{10} - \frac{3}{2}) e^{-2x}\\
&= \frac{5}{2}e^{10-2x} - \frac{3}{2} e^{-2x}
\end{align*}

The combined particular solution would then be:
$$
y(x) =
\begin{cases}
\frac{5}{2} - \frac{3}{2} e^{-2 \cdot x}, & 0 \leq x \leq 5 \\
\frac{5}{2}e^{10-2x} - \frac{3}{2} e^{-2x}, & x > 5
\end{cases}
$$
