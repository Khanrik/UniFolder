```python
# Masses in kg
m1 = 0.03
m2 = 0.018
m3 = 0.03
m4 = 0.018
m5 = 0.03

# Radii of masses in cm
d1 = 5
d2 = 10
d3 = 15
d4 = 20
d5 = 25

# Use formula tau = r * m * g * sin(theta).
# let g not be used, to have the unit be kg*cm.
# let theta = pi/2 to get the max torque.

tau = m1*d1 + m2*d2 + m3 * d3 + m4 * d4 + m5 * d5 

print("tau =", tau, "kg*cm")
```
