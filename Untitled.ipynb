```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
omega = np.linspace(-4*np.pi, 4*np.pi, 1000)
x = 1 / (1 - 1/2 * np.exp(1j * omega))
ax.plot(omega, x)
```

```python
fig, ax = plt.subplots()
omega = np.linspace(-4*np.pi, 4*np.pi, 1000)
x = 1 / (1 - 1/2 * np.exp(1j * omega))
ax.plot(omega, np.abs(x))
```

```python
fig, ax = plt.subplots()
omega = np.linspace(-4*np.pi, 4*np.pi, 1000)
x = 1 / (1 - 1/2 * np.exp(1j * omega))
ax.plot(omega, np.angle(x))
```
