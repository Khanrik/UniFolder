## Problem *9.1

### (a)

We insert a helping term $v[n] = x[n] + \frac{1}{3}v[n-1]$

![Opg9.1a](Opg9.1a.png)

The difference equation then becomes

\begin{align*}
y[n] &= 3(2x[n] + v[n]) + 6v[n-1]\\
&= 6x[n] + 3v[n] + 6v[n-1]
\end{align*}

To get rid of the $v$ terms, we find the transfer function $\frac{Y(z)}{X(z)}$ as follows

\begin{align*}
y[n] &= 3(2x[n] + v[n]) + 6v[n-1]\\
&= 6x[n] + 3v[n] + 6v[n-1]
\end{align*}

\begin{align*}
V(z) &= X(z) + \frac{1}{3} z^{-1} V(z)\\
\frac{V(z)}{X(z)} &= \frac{1}{1 - \frac{1}{3} z^{-1}}\\
X(z) &= \frac{V(z)}{1 - \frac{1}{3} z^{-1}}
\end{align*}

\begin{align*}
Y(z) &= 6X(z) + 3V(z) + 6z^{-1} V(z)\\
Y(z) &= \frac{6V(z)}{1 - \frac{1}{3} z^{-1}} + 3V(z) + 6z^{-1} V(z)\\
\frac{Y(z)}{X(z)} &= \frac{6 \cdot \frac{1}{1 - \frac{1}{3} z^{-1}}}{1 - \frac{1}{3} z^{-1}} + 3 \frac{1}{1 - \frac{1}{3} z^{-1}} + 6z^{-1} \frac{1}{1 - \frac{1}{3} z^{-1}}\\
&= \frac{6 - \frac{6}{3} z^{-1}}{1 - \frac{1}{3} z^{-1}} + \frac{3}{1 - \frac{1}{3} z^{-1}} + \frac{6z^{-1}}{1 - \frac{1}{3} z^{-1}}\\
&= \frac{9 + 4 z^{-1}}{1 - \frac{1}{3} z^{-1}}\\
&\Updownarrow\\
y[n] &= 9x[n] + 4x[n-1] + \frac{1}{3} y[n-1]
\end{align*}

### (b)

To get it down to an easily z-transformable form, and then doing the inverse transform.

\begin{align*}
H(z) = \frac{Y(z)}{X(z)} &= \frac{9 + 4 z^{-1}}{1 - \frac{1}{3} z^{-1}}\\
&= \frac{-12 + 4z^{-1}}{1 - \frac{1}{3} z^{-1}} + \frac{21}{1 - \frac{1}{3} z^{-1}}\\
&= \frac{-12 \left( 1 - \frac{1}{3} z^{-1} \right)}{1 - \frac{1}{3} z^{-1}} + 21 \frac{1}{1 - \frac{1}{3} z^{-1}}\\
&= -12 + 21 \frac{1}{1 - \frac{1}{3} z^{-1}}\\
\Updownarrow\\
h[n] &= -12 \delta[n] + 21 \cdot \left( \frac{1}{3} \right)^{n} u[n]
\end{align*}
