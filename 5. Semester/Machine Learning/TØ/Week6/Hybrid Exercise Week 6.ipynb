Consider a training data set of inputs $\{t_1 = −1, t_2 = 1, t_3 = 2\}$ and the corresponding targets or outputs $\{x_1 = 1, x_2 = 1.5, x_3 = 3\}$. The goal is to first fit an appropriate linear regression model to the above data using the least squares approach and then use that model to predict the value of x for a new value of t.



## 1) (Score: 25 %) Considering the following choice for the basis functions $\Phi(t) = \{\Phi_0(t), \Phi_1(t)\} = \{1, t\}$ that leads to the simplest linear model for regression $x(t, w) = w_0 + w_1 t$, write down the corresponding normal equations for the least squares problem.

Equation 3.15 in the book

## 2) (Score: 25 %) Solve the normal equations to obtain the parameters or the weights (w0 and w1) of the linear model. Hint: The inverse of a 2 × 2 matrix is 
$$
A^{-1} = 
\begin{bmatrix}
a & b \\ c & d 
\end{bmatrix}^{-1} = \frac{1}{ad - bc} 
\begin{bmatrix}
d & -b \\ -c & a 
\end{bmatrix}.
$$

```python
t = np.array([-1, 1, 2])[:, np.newaxis]
x = np.array([1, 1.5, 3])[:, np.newaxis]
N = len(t)
M = 2
bias = np.ones((N,1))
Phi = np.hstack((bias, t))
w_ML = np.linalg.inv(Phi.T @ Phi) @ Phi.T @ x
w_ML
```

## 3) (Score: 25 %) Plot the linear model function x(t, w) = w0 + w1t, obtained in the previous step, in Fig. 1. You can assess your model by its ability to fit the training data i.e., how close it is to the training data.

```python
tspace = np.linspace(min(t), max(t))
plt.scatter(t, x, label="target", marker="^")
plt.plot(tspace, w_ML[0,0] + w_ML[1,0]*tspace, label="linear model0")
plt.legend()
```

## 4) (Score: 25 %) Using the linear model that is estimated in the previous steps, predict the value of x for t = 0, and for t = 3. Plot the corresponding points in Fig. 1 as crosses.

```python
t_new = np.array([0,3])[:,np.newaxis]
plt.plot(t,x,'x')
plt.plot(t_new,w_ML[0,0] + w_ML[1,0]*t_new,'o')
```
