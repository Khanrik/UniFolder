## Problem 1 {-}
Consider the third-order differential equation

\begin{equation}
\tag{1}
y^{(3)} + 3y'' - 4y = x^{2} + 1
\end{equation}

### 1. {-}
Show that $y_{1}(x) = e^{x}$, $y_{2}(x) = e^{−2x}$, and $y_{3}(x) = xe^{−2x}$ are all solutions of the corresponding homogeneous equation.

The corresponding homogenous equation to (1) would be
$$
y^{(3)} + 3y'' - 4y = 0
$$

To confirm that the three functions are solutions to the homogenous equation, they and their derivatives can be inserted into the equation. Starting with $y_{1}$, its relevant derivatives are
$$
y_{1}^{(3)}(x) = e^{x}, \qquad y_{1}''(x) = e^{x}, \qquad y_{1}(x) = e^{x}
$$

Inserting these into the equation gives
$$
e^{x} + 3 e^{x} - 4 e^{x} = 4e^{x} - 4 e^{x} = 0 \qquad \text{(true statment)}
$$

The derivatives of $y_{2}$ are
$$
y_{2}^{(3)}(x) = -8e^{-2x}, \qquad y_{2}''(x) = 4e^{-2x}, \qquad y_{2}(x) = e^{-2x}
$$

Inserting these into the equation gives
$$
-8e^{-2x} + 3 (4e^{-2x}) - 4 (e^{-2x}) = -12 e^{-2x} + 12e^{-2x} = 0 \qquad \text{(true statment)}
$$

Lastly, the derivatives of $y_{3}$ are
$$
y_{3}^{(3)}(x) = 12e^{-2 x} - 8 x e^{-2 x}, \qquad y_{3}''(x) = 4 x e^{-2 x} - 4 e^{-2 x}, \qquad y_{3}(x) = xe^{-2x}
$$

Inserting these into the equation gives
$$
(12e^{-2 x} - 8 x e^{-2 x}) + 3 (4 x e^{-2 x} - 4 e^{-2 x}) - 4 (xe^{-2x}) = 12e^{-2 x} - 12e^{-2 x} - 12xe^{-2 x} + 12xe^{-2 x} = 0 \qquad \text{(true statment)}
$$

Since all the functions can be inserted into the homogenous equation without making it a false statement, they must all be solutions to the homogenous equation.

### 2. {-}
Verify by computation of the Wronskian that $y_{1}(x)$, $y_{2}(x)$, and $y_{3}(x)$ form a linearly independent set.

The wronskian to the solution is set up as 
$$
W =
\begin{bmatrix}
y_{1}(x) & y_{2}(x) & y_{3}(x) \\
{y_{1}}'(x) & {y_{2}}'(x) & {y_{3}}'(x) \\
{y_{1}}''(x) & {y_{2}}''(x) & {y_{3}}''(x)
\end{bmatrix}
$$

Inserting the values of the different solutions and their respective derivatives (found with sympy in python) into the wronskian gives

````python
x = sp.symbols('x')
y1 = sp.exp(x)
y2 = sp.exp(-2*x)
y3 = x*sp.exp(-2*x)

W = sp.Matrix([ [y1, y2, y3],
                [sp.diff(y1, x), sp.diff(y2, x), sp.diff(y3, x)],
                [sp.diff(y1, x, x), sp.diff(y2, x, x), sp.diff(y3, x, x)]])
W
````

By finding the determinant of this wronskian it can be determined whether the solutions which the wronskian is constructed from are linearly independent or not. Using sympy's determinant function, the wronskian is computed to be

````python
sp.det(W)
````

Since this value is different from 0, the solutions are linearly independent.

## Problem 2  {-}
Let a second-order differential equation be given as

\begin{equation}
\tag{2}
x'' + \beta x' + 2x = 2 \cos(t)
\end{equation}

### 1. {-}
Solve the differential equation assuming $\beta = 3$, $x(0) = 1$, and $x'(0) = 0$. Plot the solution for $t \geq 0$.

To solve the differential equation, the general solution has to be found. With the given value of $\beta = 3$, the auxiliary of the complementary homogeneous equation becomes
$$
r^{2} + 3r + 2 = 0
$$

This expression can be factorized as follows
$$
(r + 1) (r + 2) = 0
$$

which means that it has the roots $r = -1$ and $r = -2$, giving the complementary equation
$$
x_{c} = c_{1} e^{-t} + c_{2}e^{-2t}
$$

Since the right hand side of the original equation is a sinosoidal function, the trial used to find the particular solution will be of the form
$$
x_{p} = A \cos(t) + B \sin(t)
$$

The derivatives of this particular solution would be
$$
x_{p}' = -A \sin(t) + B \cos(t)
$$
$$
x_{p}'' = -A \cos(t) - B \sin(t)
$$

Inserting these expressions into the original differential equation gives

\begin{align*}
-A \cos(t) - B \sin(t) + 3 (-A \sin(t) + B \cos(t)) + 2(A \cos(t) + B \sin(t)) &= 2 \cos(t)\\
-A \cos(t) - B \sin(t) - 3A \sin(t) + 3B \cos(t) + 2A \cos(t) + 2B \sin(t) &= 2 \cos(t)\\
A \cos(t) + 3B \cos(t) + B \sin(t) - 3A \sin(t) &= 2 \cos(t)\\
(A + 3B) \cos(t) + (B - 3A) \sin(t) &= 2\cos(t)
\end{align*}

This will be true for all $t$ provided that the cosine and sine terms on the two sides of this equation agree. It therefore sufﬁces for A and A to satisfy the two linear equations

\begin{align*}
A + 3B &= 2\\
B - 3A &= 0
\end{align*}

Solving this by substitution gives

\begin{align*}
B &= 3A &\text{(isolating B in equation 2)}\\
A + 3 (3A) &= 2 &\text{(substituting into equation 1)}\\
10A &= 2\\
A &= \frac{1}{5}\\
B &= 3 \cdot \frac{1}{5} &\text{(substituting back into A)}\\
B &= \frac{3}{5}
\end{align*}

This means that the general solution for the differential equation and its derivative is

\begin{align*}
x(t) &= c_{1} e^{-t} + c_{2}e^{-2t} + \frac{1}{5} \cos(t) + \frac{3}{5} \sin(t)\\
x'(t) &= -c_{1}e^{-t} - 2c_{2}e^{-2t} - \frac{1}{5} \sin(t) + \frac{3}{5} \cos(t)
\end{align*}

Inserting the inital value for the general solution gives

\begin{align*}
c_{1} e^{-0} + c_{2}e^{-2 \cdot 0} + \frac{1}{5} \cos(0) + \frac{3}{5} \sin(0) &= 1\\
c_{1} + c_{2} + \frac{1}{5} + 0 &= 1\\
c_{1} &= 1 - \frac{1}{5} - c_{2}\\
c_{1} &= \frac{4}{5} - c_{2}
\end{align*}

Inserting the found value of $c_{1}$ and the inital value into the derivative gives

\begin{align*}
-\left( \frac{4}{5} - c_{2} \right) e^{-0} - 2c_{2}e^{-2 \cdot 0} - \frac{1}{5} \sin(0) + \frac{3}{5} \cos(0) &= 0\\
-\frac{4}{5} + c_{2} - 2c_{2} - 0 + \frac{3}{5} &= 0\\
-c_{2} - \frac{1}{5} &= 0\\
c_{2} &= - \frac{1}{5}
\end{align*}

Inserting back into $c_{1}$ gives
$$
c_{1} = \frac{4}{5} - (- \frac{1}{5}) = \frac{5}{5} = 1
$$

The particular solution of the differential equation then becomes
$$
x(t) = e^{-t} - \frac{1}{5}e^{-2t} + \frac{1}{5} \cos(t) + \frac{3}{5} \sin(t)
$$

Plotting the solution results in the following

````python
#Define ODE
def sol(t):
    return np.exp(-t) - 1/5 * np.exp(-2*t) + 1/5 * np.cos(t) + 3/5 * np.sin(t)

#Define boundaries of x and y in the slope field plot
minx = 0.
maxx = 20.
miny = -1
maxy = 1.25

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.set_xlabel("t")
ax.set_ylabel("x")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()

t_sol = np.linspace(minx, maxx, 1000)
x_sol = sol(t_sol)
ax.plot(t_sol, x_sol)
````

### 2. {-}
Account for the transient and stationary behaviour of the solution.

It can be seen that the graph starts at $x=1$ but never reaches $x=-1$ before it becomes stationary as a periodic sinusoid defined by $\frac{1}{5} \cos(t) + \frac{3}{5} \sin(t)$. The reason is that there are two transient terms in the expression of the solution, which are $e^{-t}$ and $- \frac{1}{5}e^{-2t}$. Out of these two, $e^{-t}$ is the largest one and is the one giving the value of 1 at $t=0$, since the other transient term cancels out with the cosine term. The first trough appears right before $t=5$. When looking at the transient terms, they already start to approach very small sizes at this value of $t$
$$
e^{-5} \approx 0.0067, \qquad - \frac{1}{5}e^{-2 \cdot 5} \approx - 0.0000091
$$

Their transient property is therefore already showing making the solution approach the stationary sinusoid almost immediately.

### 3. {-}
What happens if $\beta$ is increased?

To know what happens, it has to be analyzed what $\beta$ contributes. When increasing $\beta$, it changes the root of the auxiliary of the complementary homogeneous equation. How it changes can be seen by writing the quadratic formula with a symbolic $\beta$
$$
r = \frac{- \beta \pm \sqrt{ \beta^{2} - 4 \cdot 1 \cdot 2 }}{2 \cdot 1}
$$

Since $\beta$ is the only changing component, the expression can be taken to the extreme to see the behavior as $\beta$ increases. As $\beta$ approaches $\infty$ it dominates both the other term in the squareroot as well as the denominator. The expression can therefore be approximated to
$$
r \approx - \beta \pm \beta \qquad (\beta \text{ very large})
$$

This means that as $\beta$ approaches $\infty$ the two roots approach $-\infty$ and 0. Since the roots are different negative values at $\beta = 3$, neither root will ever become positive nor be equal to the other as $\beta$ increases. The result of this is that as $\beta$ increases, the complementary homogeneous equation approaches 
$$
x_{c} = c_{1} e^{0} + c_{2}e^{-\infty} = c_1
$$

The effect that increasing $\beta$ has on the particular solution part can be boiled down to the following expressions

\begin{align*}
A + \beta B &= 2\\
B - \beta A &= 0
\end{align*}

This is due to the same calculations made in problem 2.1 where most terms cancel out except those containing $\beta$. Isolating the factors gives

\begin{align*}
B &= \beta A &\text{(isolating B in equation 2)}\\
A + \beta (\beta A) &= 2 &\text{(substituting into equation 1)}\\
(\beta^{2} + 1) A &= 2\\
A &= \frac{2}{\beta^{2} + 1}\\
B &= \beta \cdot \frac{2}{\beta^{2} + 1} &\text{(substituting back into A)}\\
B &= \frac{2 \beta}{\beta^{2} + 1}
\end{align*}

Just as before, since the problem asks what happens when $\beta$ increases, the extreme case can be analyzed. When $\beta$ approaches $\infty$ both factors of the particular solution would approach 0 as the denominator of both would dominate. This leads to the particular solution being
$$
x_{p} = \frac{2}{\infty^{2} + 1} \cos(t) + \frac{2 \infty}{\infty^{2} + 1} \sin(t) = 0 \cos(t) + 0 \sin(t)
$$

This means that the amplitude of the sinusoid will decrease more and approach a constant value of 0 as $\beta$ increases. Combining everything the general solution would end up approaching
$$
x = c_{1}
$$

In other words, as $\beta$ increases one of the transient terms dies out faster while the other approaches a constant value and the sinusoidal parts will decrease in amplitude. As $\beta$ approaches $\infty$ the solution will no longer be transient, but just stationary at a constant value instead of a sinusoid.

### 2. {-}
In a physical system it is impossible to measure the parameters precisely. Assume that the spring constant is instead $k = 1 − \epsilon^{2}$ where $\epsilon$ is a small number.

Show that in this case, with the same initial values as above, that the solution is $x_{2}(t) = e^{−t} \cosh(\epsilon t)$.

The derivatives of this solution are not as straight forward, so instead the solution of the intial value problem is found and then compared to the given solution. First the general solution therefore has to be found. Since the differential equation is homogeneous this can be done based on the roots of the auxiliary equation which is
$$
r^{2} + 2r + 1 - \epsilon^{2} = 0
$$

The roots of this can be found with the quadratic formula

\begin{align*}
r &= \frac{- 2 \pm \sqrt{ 2^{2} - 4 \cdot 1 \cdot (1-\epsilon^{2}) }}{2 \cdot 1}\\
&= \frac{- 2 \pm \sqrt{ 4 - 4 + 4\epsilon^{2} }}{2}\\
&= \frac{- 2 \pm 2\epsilon}{2}\\
&= - 1 \pm \epsilon
\end{align*}

Assuming that $\epsilon$ is not complex, the general solution of the differential equation and its derivative will be

\begin{align*}
x &= c_{1} e^{(- 1 + \epsilon) t} + c_{2} e^{(- 1 - \epsilon) t}\\
x' &= (- 1 + \epsilon)c_{1} e^{(- 1 + \epsilon) t} + (- 1 - \epsilon) c_{2} e^{(- 1 - \epsilon) t}
\end{align*}

Inserting the inital value for the general solution gives

\begin{align*}
c_{1} e^{(- 1 + \epsilon) 0} + c_{2} e^{(- 1 - \epsilon) 0} &= 1\\
c_{1} + c_{2} &= 1\\
c_{1} &= 1 - c_{2}
\end{align*}

Inserting the found value of $c_{1}$ and the inital value into the derivative gives

\begin{align*}
(- 1 + \epsilon) (1 - c_{2}) e^{(- 1 + \epsilon) 0} + (- 1 - \epsilon) c_{2} e^{(- 1 - \epsilon) 0} &= -1\\
- 1 + c_{2} + \epsilon - \epsilon c_{2} - c_{2} - \epsilon c_{2}  &= -1\\
- 1 + \epsilon - 2 \epsilon c_{2} &= -1
\end{align*}

For the equality to be satisfied in this expression all terms on the left side must be equal to 0 except the -1 term. The value of $c_{2}$ can therefore be found as follows

\begin{align*}
\epsilon - 2 \epsilon c_{2} &= 0\\
- 2 \epsilon c_{2} &= - \epsilon\\
c_{2} &= \frac{1}{2}
\end{align*}

Inserting this back into $c_{1}$ gives
$$
c_{1} = 1 - \frac{1}{2} = \frac{1}{2}
$$

The solution to the initial value problem with the given spring constant can therefore be expressed as

\begin{align*}
x_{2}(t) &= \frac{1}{2} e^{(- 1 + \epsilon) t} + \frac{1}{2} e^{(- 1 - \epsilon) t}\\
x_{2}(t) &= \frac{e^{(- 1 + \epsilon) t} + e^{(- 1 - \epsilon) t}}{2}\\
x_{2}(t) &= \frac{e^{- t} \cdot e^{\epsilon t} + e^{- t} \cdot e^{-\epsilon t}}{2}\\
x_{2}(t) &= e^{- t} \frac{e^{\epsilon t} + e^{-\epsilon t}}{2}\\
x_{2}(t) &= e^{- t} \cosh(\epsilon t) &\text{using definition:} \cosh(x) = \frac{e^{x} + e^{-x}}{2}
\end{align*}

The definition used above is taken from [Wikipedia](https://en.wikipedia.org/wiki/Hyperbolic_functions#Exponential_definitions). Since the particular solution matches the one, which the exercise suggested to be the solution, it has been proven that it is indeed a valid solution for the given initial value problem.


### 3. {-}
Comment on the difference(s) between the two solutions and the implication(s).

The two solutions are very similar since they both contain the factor $e^{-t}$. The difference is that $x_{2}(t)$ is multiplied by a hyperbolic cosine factor. The hyperbolic cosine function looks as follows

````python
#Define ODE
def sol(t):
    return np.cosh(t)

#Define boundaries of x and y in the slope field plot
minx = -4.
maxx = 4.
miny = 0
maxy = 8

fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.set_xlabel("t")
ax.set_ylabel("x")
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
ax.grid()

t_sol = np.linspace(minx, maxx, 1000)
x_sol = sol(t_sol)
ax.plot(t_sol, x_sol, label="x = cosh(t)")
ax.hlines(1, minx, maxx, colors='r', linestyles='dashed', label='x = 1')
ax.legend(fontsize=14)
````

The smaller the input value of the $\cosh$ function is, the closer the output is to 1. Since the value of $\epsilon$ is very small, the $\cosh(\epsilon t)$ factor will be close to 1 at lower values of $t$, which make the output of the solution slightly larger, essentialy working as an error factor for the solution. The reason for needing such an error function is that, as the problem describes, measuring physical parameters can be imprecise. Looking at the critical damping coefficient for the problem of $x_{1}(t)$, it is

$$
c_{cr} = \sqrt{4 \cdot 1 \cdot 1} = 2
$$

Since the damping coefficient is 2 and 2 = 2, the solution to $x_{1}(t)$ is critically damped. With the slightest change to the spring constant however, it can quickly be over- or underdamped. The critical damping coefficient for the problem of $x_{2}(t)$ is

$$
c_{cr} = \sqrt{4 \cdot (1 - \epsilon^{2}) \cdot 1} = \sqrt{4 - 4\epsilon^{2}} \qquad -1 \leq \epsilon \leq 1
$$

This adds an extra constraint to $\epsilon$ since the critical damping coefficient should be a real number. It is seen that though $\epsilon$ does not change much since it is a small number, it does slightly decrease the critical damping coefficient, which makes the entire system overdamped. Using the solution $x_{2}(t)$ therefore enables you to account for the cases where the mass-dashpot-spring system does not behave like it ideally should.

## Problem 4 {-}
Consider the initial value problem

\begin{equation}
\tag{4}
y' = 3 \cos(y - 5x), \qquad y(0) = 0
\end{equation}

### 1. {-}
Use the Runge-Kutta-4 algorithm to approximate the solution in the range $0 \leq x \leq 4$. Plot the approximate solution.

The Runge-Kutta-4 algorithm implemented in python as follows:

```py
def runge_kutta(func, x0, y0, h, x_end):
    Y = [y0]
    X = [x0]
    n = 0
    while X[-1] < x_end:
        k1 = func(X[n], Y[n])
        k2 = func(X[n] + 1/2 * h, Y[n] + 1/2 * h * k1)
        k3 = func(X[n] + 1/2 * h, Y[n] + 1/2 * h * k2)
        k4 = func(X[n] + h, Y[n] + h * k3)
        Y.append(Y[n] + (h/6) * (k1 + 2*k2 + 2*k3 + k4))
        X.append(X[n] + h)
        n += 1
    return X, Y
```

The input given to the function was then the differential equation, initial conditions, an end value of $x=4$, and a step size of $h = 0.0001$. This gave the approximations of various x- and y-values of the solution curve. Below is a plot of the approximated solution curve.

````python
def df(x, y):
    return 3 * np.cos(y - 5*x)

def runge_kutta(func, x0, y0, h, x_end):
    Y = [y0]
    X = [x0]
    n = 0
    while X[-1] < x_end:
        k1 = func(X[n], Y[n])
        k2 = func(X[n] + 1/2 * h, Y[n] + 1/2 * h * k1)
        k3 = func(X[n] + 1/2 * h, Y[n] + 1/2 * h * k2)
        k4 = func(X[n] + h, Y[n] + h * k3)
        Y.append(Y[n] + (h/6) * (k1 + 2*k2 + 2*k3 + k4))
        X.append(X[n] + h)
        n += 1
    return X, Y

x0 = y0 = 0
x_end = 4
h = 0.0001

x_solution, y_solution = runge_kutta(df, x0, y0, h, x_end)

fig, ax = plt.subplots()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(x_solution, y_solution)
````
