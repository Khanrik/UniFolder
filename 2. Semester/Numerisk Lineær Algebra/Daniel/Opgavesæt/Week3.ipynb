# Opg 3.1



```python
s = 2
t = 3
A = np.array([[2, 1, -1], 
              [0, 2,  5]])
B = np.array([[1, -1,  0],
              [0,  1, -1]])
```

## (a)

Denne er tilladt, da det blot er et scalar-matrix produkt:

$$sB = \begin{bmatrix} 2 & -2 & 0 \\ 0 & 2 & -2 \end{bmatrix}$$

## (b)

Denne er også tilladt da matricerne $A$ og $tB$ har samme shape:

$$A + tB = \begin{bmatrix} 5 & -2 & -1 \\ 0 & 5 & 2 \end{bmatrix}$$

## (c)

Denne er tilladt da vi blot transponerer matricen:

$$A^T = \begin{bmatrix} 2 & 0 \\ 1 & 2 \\ -1 & 5\end{bmatrix}$$

## (d)

Denne er ikke tilladt da $A^T$ og $sB$ ikke har samme shape.

## (e)

Denne er tilladt da de to matricer har samme shape.

$$sA^T - tB^T$$

```python
s*A.T - t*B.T
```

## (f)

Dette går ikke fordi den første har en forskellig mængde søjler fra den mængde rækker den anden har.

$$AB$$

## (g)

Same

$$BA$$

## (h)

Dette kan man godt:

$$A^T B$$

```python
A.T @ B
```

## (i)

Dette må man også gerne:

$$sAB^T$$

```python
s * A @ B.T
```

## (j)

Dette kan man godt:

$$BA^TBt$$

```python
B @ A.T @ B * t
```
