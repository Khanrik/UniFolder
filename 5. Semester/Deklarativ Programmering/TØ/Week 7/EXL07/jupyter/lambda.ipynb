## (1) Implement the fibonacci number using the Y-combinator. So that:
```ocaml
eval_int_lazy (yfib ^ " @ 3")
```
returns 3.

````ocaml
let yfib = "(fun g. 
(fun y. g @ (y @ y)) @ 
(fun z. g @ (z @ z))) @
(fun f x. 
  ? @ (= @ x @ 0) 
  @ 1 
  @ ( 
    ? @ ( = @ x @ 1) 
    @ 1 
    @ ( + @ (f @ ( - @ x @ 1)) @ (f @ ( - @ x @ 2)))))"
````

````ocaml
eval_int_lazy (yfib ^ " @ 3")
````

## (2) Implement the wellknown even and odd in the same way

````ocaml
let even = "(fun g. 
(fun y. g @ (y @ y)) @ 
(fun z. g @ (z @ z))) @
(fun f x. 
  ? @ (= @ x @ 0) 
  @ true
  @ ( 
    ? @ ( = @ x @ 1) 
    @ false
    @ ( f @ (- @ x @ 2))))"
````

````ocaml
eval_int_lazy (even ^ " @ 2")
````

````ocaml
let odd = "(fun g. 
(fun y. g @ (y @ y)) @ 
(fun z. g @ (z @ z))) @
(fun f x. 
  ? @ (= @ x @ 0) 
  @ false
  @ ( 
    ? @ ( = @ x @ 1) 
    @ true
    @ ( f @ (- @ x @ 2))))"
````

````ocaml
eval_int_lazy (odd ^ " @ 2")
````

## (3) Correct the substitution function from the lecture to deal correctly with free/bound variables.

Hint: You need to implement functions to compute free/bound variables for this.

````ocaml
let rec free_vars = function
  | Var x -> [x]
  | Con (_, args) -> List.flatten (List.map free_vars args)
  | Abs (x, body) -> List.filter (fun y -> y <> x) (free_vars body)
  | App (t1, t2) -> free_vars t1 @ free_vars t2

let fresh_var =
  let counter = ref 0 in
  fun base ->
    incr counter;
    base ^ "'" ^ string_of_int !counter

let rec subst x t = function
  | Var y as t' ->
      if y = x then t else t'
  | Con (c, args) ->
      Con (c, map (subst x t) args)
  | App (t1, t2) ->
      App (subst x t t1, subst x t t2)
  | Abs (y, b) as t' ->
      if y = x then
        t'
      else if mem y (free_vars t) then
        let y' = fresh_var y in
        let b' = subst y (Var y') b in
        Abs (y', subst x t b')
      else
        Abs (y, subst x t b)
````
