 & = \text{det}\left( \begin{bmatrix}

 & = (6 - \lambda) (2 - \lambda) (- 1 - \lambda)
\end{align*}
$$

$$
(6 - 6) (2 - 6) (- 1 - 6) = 0 \cdot (-4) \cdot (-7) = 0
$$

$$
(6 - 2) (2 - 2) (- 1 - 2) = 4 \cdot 0 \cdot (-3) = 0
$$

$$
(6 - (-1)) (2 - (-1)) (- 1 - (-1)) = 7 \cdot 3 \cdot 0 = 0
$$

For en vilkårlig diagonalmatrix ville man kunne stille det karakteristiske polynomium op på formen:
$$
(\sigma_{0} - \lambda) (\sigma_{1} - \lambda) (\sigma_{2} - \lambda)
$$
Derfor vil alle diagoalværdier være egenværdierne.

Den er diagonliserbar, da den er kvadratisk ($\mathbb{R}^{3 \times 3}$) og har 3 egenværdier og vektorer, hvilket betyder at den kan stilles på formen $A = V \Lambda V^{-1}$.
