### (a)

```python
u = np.array([1,-1,1])[:, np.newaxis]
v = np.array([-1,-2,1])[:, np.newaxis]
weight = np.array([7,2,3])[:, np.newaxis]

print(np.vdot(weight * u, v))
```

Siden det indre produkt er 0 vil u og v være ortogonal.

### (b)

```python
w = np.array([1,0,0])[:, np.newaxis]

w_pr_u = np.vdot(w, weight * u) / np.linalg.norm(u)**2 * u
w_pr_v = np.vdot(w, weight * v) / np.linalg.norm(v)**2 * v

print("pr_u (w) =\n", w_pr_u, "\n")
print("pr_v (w) =\n", w_pr_v)
```

## Opgave 7.2

### (a)

```python
ligning_a = np.array([  [1, 1, 1],
                        [0, 1, 1],
                        [0, 0, 1]])

# Overflødigt
# # R0 -> R0 - R1
# ligning_a[0, :] -= ligning_a[1, :]

# # R1 -> R1 - R2
# ligning_a[1, :] -= ligning_a[2, :]

print(ligning_a)
```

Da alle rækker har pivotelementer er den lineær uafhængig.

### (b)

```python
ligning_b = np.array([  [1, 1, 0],
                        [1, 0, 1],
                        [0, 1, 1]], dtype=float)

# R1 -> R1 - R0
ligning_b[1, :] -= ligning_b[0, :]

# R1 -> R1 * 1/-1
ligning_b[1, :] *= ligning_b[1, 1]

# R2 -> R2 - R1
ligning_b[2, :] -= ligning_b[1, :]

# R2 -> R2 * 1/2
ligning_b[2, :] *= 1 / ligning_b[2, 2]

print(ligning_b)
```

Da alle rækker har pivotelementer er den lineær uafhængig.

### (c)

```python
ligning_c = np.array([  [1, -1,  0],
                        [-1, 0,  1],
                        [0,  1, -1]])

# R1 <-> R2
ligning_c[[1, 2], :] = ligning_c[[2, 1], :]

# R2 -> R2 + R0
ligning_c[2, :] += ligning_c[0, :]

# R2 -> R2 + R1
ligning_c[2, :] += ligning_c[1, :]

print(ligning_c)
```

Da den nederste række ikke har et pivotelement kan den ikke være lineær uafhægnig.

### (d)

```python
ligning_d = np.array([  [1     , 1j    ,-1 + 1j],
                        [1j    , 1 - 1j, 1     ],
                        [1 + 1j, 1     , 1j    ]])

# R1 -> R1 - R2
ligning_d[1, :] -= ligning_d[2, :]

# R1 -> R1 + R0
ligning_d[1, :] += ligning_d[0, :]

# R2 <-> R1
ligning_d[[1, 2], :] = ligning_d[[2, 1], :]

print(ligning_d)
```

Da den nederste række ikke har et pivotelement kan den ikke være lineær uafhægnig.

## Opgave 7.3

### (a)

```python
# Ganger alle indgange sammen, sætter i anden, 
# summerer og tager kvadratrod
def frob_norm(M):
    sum = 0
    for row in M:
        for element in row:
            sum += element**2
    return np.sqrt(sum)
```

```python
A = np.array([  [6, 0 ],
                [0, -3]])
B = np.array([  [ 1,  2],
                [-1, -2]])

print("||A||_F =\n", frob_norm(A), "\n")
print("||B||_F =\n", frob_norm(B))
```

```python
print("<A, B> =", np.vdot(A, B))
```

```python
print("<A, A> <= ", np.vdot(A, A))
```
