### Problem 1
$\frac{dy}{dx} + 2xy = 0$

Standard form defines $P(x) = 2x$ and $Q(x) = 0$
\begin{align*}
v(x) &= e^{\int 2x \, dx}\\
&= e^{x^{2}}\\
\end{align*}

\begin{align*}
y &= \frac{1}{e^{x^{2}}} \int e^{x^{2}} \cdot 0 \, dx\\
&= e^{-x^{2}} \cdot C\\
&= C e^{-x^{2}}
\end{align*}

### Problem 19
$\frac{dy}{dx} = y e^{x}, \qquad y(0) = 2e$

It can be solved as a separable differential equation
\begin{align*}
\frac{dy}{dx} &= y e^{x}\\
\frac{1}{y} dy &= e^{x} dx\\
\int \frac{1}{y} \, dy &= \int e^{x} \, dx\\
\ln(y) &= e^{x} + C_{1}\\
y &= e^{e^{x} + C_{1}}\\
y &= e^{e \cdot x} + e^{C_{1}}\\
y &= e^{e^{x}} + C_{2}
\end{align*}

The inital value can then be inserted giving
\begin{align*}
2e &= e^{e^{0}} \cdot C_{2}\\
2e &= e^{1} \cdot C_{2}\\
C_{2} &= \frac{2e}{e}\\
C_{2} &= e
\end{align*}
