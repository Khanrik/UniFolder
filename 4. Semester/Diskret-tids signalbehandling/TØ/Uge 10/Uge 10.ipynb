```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

## Problem *7.1

### (a) FIND NOGET BEGRUNDELSE FOR LAPLACE TIL CTFT

$$
x_{c}(t) = 5 e^{ -10 t } \sin(20 \pi t) u(t)
$$

The original signal is transformed into the following using a Laplace transformation.

$$
\begin{align*}
X_{c}(s) &= 5 \frac{20\pi}{(s + 10)^{2} + (20\pi)^{2}}\\
&= \frac{100\pi}{(s + 10)^{2} + 400\pi^{2}}\\
\end{align*}
$$

Then $s$ can be substituted with $j 2\pi F$ to get the CTFT

$$
\begin{align*}
X_{c} (j 2\pi F) &= \frac{100\pi}{(j 2\pi F)^{2} + 100 + 20(j 2\pi F) + 400\pi^{2}}
\end{align*}
$$

### (b)

```python
num = [100 * np.pi]
den = [1, 20, 100 + 400 * np.pi**2]

w = 2 * np.pi * np.linspace(-50, 50, 1000)  # rad/s
w, H = signal.freqs(num, den, worN=w)

plt.figure(figsize=(12,5))
plt.plot(w/(2*np.pi), np.abs(H))
plt.title('Magnitude Response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
```

```python
plt.plot(w/(2*np.pi), np.angle(H))
plt.title('Phase Response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [radians]')
plt.grid()
plt.tight_layout()
plt.show()
```

### (c)

```python
# Parameters
fs = 1000  # Sampling frequency in Hz
T = 1      # Duration in seconds
t = np.linspace(0, T, int(fs*T), endpoint=False)

# Original signal (u(t) is 1 for t >= 0)
x = 5 * np.exp(-10 * t) * np.sin(20 * np.pi * t)

# FFT
X = np.fft.fft(x)
freqs = np.fft.fftfreq(len(x), 1/fs)

# Normalize the magnitude
X_mag = np.abs(X) / np.max(np.abs(X))

# Only plot the positive and negative frequencies in order
idx = np.argsort(freqs)
plt.figure(figsize=(12,5))
plt.plot(freqs[idx], X_mag[idx])
plt.xlim(-50, 50)  # Adjust as needed
plt.title('Normalized FFT Magnitude of $x_c(t)$')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Normalized Magnitude')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(freqs[idx], np.angle(X)[idx])
plt.xlim(-50, 50)  # Adjust as needed
plt.title('FFT Magnitude of $x_c(t)$')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
