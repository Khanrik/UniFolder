## Opgave 2.1

```python
a = np.array([0.0, 1.0])
b = np.array([1.0, -1.0])
c = np.array([-0.5, -0.8660254037844386])

a_len = np.sqrt(a[0]**2 + a[1]**2)
b_len = np.sqrt(b[0]**2 + b[1]**2)
c_len = np.sqrt(c[0]**2 + c[1]**2)

a_rad = np.arctan2(a[1], a[0])
b_rad = np.arctan2(b[1], b[0])
c_rad = np.arctan2(c[1], c[0])

a_deg = np.degrees(a_rad)
b_deg = np.degrees(b_rad)
c_deg = np.degrees(c_rad)

print(f"a = {a_len}, {a_rad}, {a_rad / np.pi} π, {a_deg}")
print(f"b = {b_len}, {b_rad}, {b_rad / np.pi} π, {b_deg}")
print(f"c = {c_len}, {c_rad}, {c_rad / np.pi} π, {c_deg}")
```

## Opgave 2.4

I hver af de følgende to tilfælder, bestem alle matricer $A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$, der opfylder de givne krav:

### (a)

$Ae_{0} = \begin{bmatrix} 4 \\ -3 \end{bmatrix}$ og $Ae_{1} = \begin{bmatrix} 2 \\ 5 \end{bmatrix}$, hvor $e_{i}$ er standardvektorerne.

```python
e0 = np.array([1.0, 0.0])[:, np.newaxis]
e1 = np.array([0.0, 1.0])[:, np.newaxis]
A = np.array([[4, 2],
              [-3, 5]])

print("Ae0 is \n", A @ e0)
print("Ae1 is \n", A @ e1)
```

### (b)

$A$ er en spejlingsmatrix med $A \begin{bmatrix} 4 \\ -3 \end{bmatrix} = \begin{bmatrix} 5 \\ 0 \end{bmatrix}$

\begin{cases}
c \cdot 4 + s \cdot (-3) = 5 \\
s \cdot 4 - c \cdot (-3) = 0
\end{cases}

\begin{align*}
s \cdot 4 - c \cdot (-3) & = 0 \\
s & = -\frac{c \cdot 3}{4}
\end{align*}

\begin{align*}
c \cdot 4 + (- \frac{c \cdot 3}{4}) \cdot (-3) & = 5 \\
c \cdot 4 + \frac{c \cdot 9}{4} & = 5 \\
c \cdot 16 + c \cdot 9 & = 20 \\
c \cdot 25 & = 20 \\
c & = \frac{4}{5}
\end{align*}

\begin{align*}
s & = - \frac{c \cdot 3}{4} \\
s & = - \frac{4}{5} \cdot \frac{3}{4} \\
s & = - \frac{12}{20} \\
s & = - \frac{3}{5}
\end{align*}

```python
c = 4/5
s = -3/5
A = np.array([[c, s],
              [s, -c]])

v = np.array([4, -3])[:, np.newaxis]

print("A [4,-3] is \n", A @ v)
```
