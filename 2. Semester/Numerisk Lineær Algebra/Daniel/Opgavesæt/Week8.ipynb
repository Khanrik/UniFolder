## (a)

Vi har at følgende skal gælde, med alle $s_i = 0$, hvis samlingen skal være uafhængig:

$$s_0 * 1 + s_1 * x + s_2 * x^3 = 0$$

Differentier 2 gange:

$$6s_2*x = 0 \Leftrightarrow s_2 = 0$$

Differentier nu kun 1 gang:

$$s_1 + s_2*x^3 = s_1 = 0$$

Indsæt nu $s_1$ og $s_2$ i den originale formel:

$$s_0 + s_1 * x + s_2 * x^3 = s_0 = 0$$

Altså må samlingen være lineært uafhængig.

## (b)



Vi har at følgende skal gælde, med alle $s_i = 0$, hvis samlingen skal være uafhængig:

$$s_0 * 1 + s_1 * e^x + s_2 * e^{-x}= 0$$

Differentier 1 og 2 gange:

1:
$$s_1 * e^x - s_2 * e^{-x}$$

2:
$$s_1 * e^x + s_2 * e^{-x}$$

Disse to ligninger kan løses, således at $s_1$ og $s_2$ er nødt til at være lig $0$

Hvis $s_1$ og $s_2$ indsættes i den originale kan det ses at $s_0$ også nødvendigvis er $0$.

# 8.5



```python
def f(a, b, c):
    return (-b+np.sqrt(b**2-4*a*c))/(2*a)
```

```python
a = 2.0
b = 2.0
c = -1.0

rel_ændring = 0.1

def rel_space(x, rel, num):
    return np.linspace(x*(1.0-rel), x*(1.0+rel), num)

av = rel_space(a, rel_ændring, 51)
bv = rel_space(b, rel_ændring, 51)
cv = rel_space(c, rel_ændring, 51)

def rel_fejl(x,y):
    return (x-y)/y

fig, ax = plt.subplots()

ax.plot(rel_fejl(av, a), f(av, b, c), label='a')
ax.plot(rel_fejl(bv, b), f(a, bv, c), label='b')
ax.plot(rel_fejl(cv, c), f(a, b, cv), label='c')
ax.legend()
```

Ovenfor kan vi se at relative ændringer i c har den største virkning på funktionsværdien.

Dette burde gerne stemme overens med hvad vores gradient siger.

For at beregne konditionstallet bruger jeg følgende formel:



```python
x = np.array([a, b, c]) [:, np.newaxis]

gradf = np.array([(-(b**2)/(np.sqrt(b**2-4*a*c)) + (2*a*c)/(np.sqrt(b**2-4*a*c)) + b)/(2*a**2),
                  (b/(np.sqrt(b**2-4*a*c))-1)/(2*a),
                  -1/(np.sqrt(b**2-4*a*c))])

x, gradf
```

Vi kan ovenfor se at den afledte mht til c er numerisk størst, altså sker den største ændring hvis der sker en fejl i c. 

Konditionstallet står nedenfor:

```python
kappa = np.linalg.norm(x) *np.linalg.norm(gradf)/np.abs(f(a, b, c))
kappa
```

# 8.6



Som følge af I1, så holder det indre produkt for alle symmetriske matricer.

Desuden er matrix operationer lineære, og dermed holder I2.

For at I3 holder, så skal følgende gælde:

Lad $x = y = e_0$. For at $e_0^T A e_0 > 0$, så skal $a_{00} > 0$.

Lad $x = y = e_1$. For at $e_1^T A e_1 > 0$, så skal $a_{11} > 0$.

Enhver vektor som ikke er proportionel med $e_0 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$, må være proportionel med:$\begin{bmatrix} t \\ 1 \end{bmatrix}$, for et eller andet $t$. (0-vektoren er ikke inkluderet).

$$\langle(t,1), (t,1) \rangle = \begin{bmatrix} t & 1 \end{bmatrix}  \begin{bmatrix} a_{00} & a_{01} \\ a_{10} & a_{11} \end{bmatrix} \begin{bmatrix} t \\ 1 \end{bmatrix}$$

$$= \begin{bmatrix} ta_{00} + a_{10} & ta_{01} + a_{11} \end{bmatrix} \begin{bmatrix} t \\ 1 \end{bmatrix}$$

$$= t(ta_{00} + a_{10}) + ta_{01} + a_{11}$$

$$= t^2 a_{00} + ta_{10} + ta_{01} + a_{11}$$

Fordi $A$ er symmetrisk kan vi skrive at:

$$= t^2 a_{00} + 2ta_{10} + a_{11}$$

Dette omskrives til:

$$= a_{00} t^2 + 2a_{10} t + a_{11}$$

Fra I3 gælder det at: 

$$a_{00} t^2 + 2a_{10} t + a_{11} > 0$$

Dette polynomium har rødderne:

$$t = \frac{-2a_{10} \pm \sqrt{(2a_{10})^2 - 4 a_{00} a_{11}}}{2a_{00}}$$

$$t = \frac{-2a_{10} \pm \sqrt{4a_{10}^2 - 4 a_{00} a_{11}}}{2a_{00}}$$

$$t = \frac{-a_{10} \pm \sqrt{a_{10}^2 - a_{00} a_{11}}}{a_{00}}$$

$a_{00}$ og $a_{11}$ er positive, dermed peger benene på polynomiet opad. Vi vil altså have ingen nulpunkter.

Hvis vi ikke skal have nogle nulpunkter, så skal vores deskriminant være negativ. Dette er den kun hvis:

$$a_{10}^2 - a_{00}a_{11} < 0 \Leftrightarrow \det(A) = a_{00}a_{11} - a_{10}^2 > 0$$

Altså er det indreprodukt beskrevet af $A$ kun et indre produkt hvis:

$A$ er symmetrisk.

$a_{00} > 0$,

$a_{11} > 0$ og

$\det(A) > 0$
