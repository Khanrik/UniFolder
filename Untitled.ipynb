```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

omega = np.linspace(-4*np.pi, 4*np.pi, 1000)
x = 1/(np.abs(1 - 1/2 * np.exp(- 1j * omega)))
x2 = 1/np.sqrt((1 - 1/2 * np.cos(omega))**2 + (1/2 * np.sin(omega))**2)

ax.plot(omega, x)
ax.plot(omega, x2)
```
