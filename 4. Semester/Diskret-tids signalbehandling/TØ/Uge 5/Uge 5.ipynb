## Problem *3.14

Given a causal system described by $y[n] = \frac{1}{2} y[n − 1] + x[n]$, compute its response to the following inputs:

### (a) $x[n] = e^{ j(\frac{\pi}{4}) n}, \quad -\infty < n < \infty$

Taking the z-transform leads to
$$
\begin{align*}
Y(z) &= \frac{1}{2} z^{-1} Y(z) + X(z)\\
Y(z) - \frac{1}{2} z^{-1} Y(z) &= X(z)\\
Y(z) \left( 1 - \frac{1}{2} z^{-1} \right) &= X(z)\\
Y(z) &= \frac{X(z)}{1 - \frac{1}{2} z^{-1}}\\
H(z) = \frac{Y(z)}{X(z)} &= \frac{1}{1 - \frac{1}{2} z^{-1}}
\end{align*}
$$

For LTI-systems, the response can be found as $y[n] = H[z] \cdot z^{n}$, where $x[n] = z^{n}$
$$
y[n] = \frac{1}{1 - \frac{1}{2} e^{ -j(\frac{\pi}{4}) n}} \cdot e^{ j(\frac{\pi}{4}) n}
$$
